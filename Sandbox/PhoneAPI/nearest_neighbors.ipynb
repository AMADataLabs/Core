```python
import pandas as pd
```

```python
numbers = pd.read_csv('../GetPhoneInfo/ALL_GETPHONE.csv')
```

```python
X = numbers.drop("label",axis = 1)
y = numbers['label']
print(X.shape, y.shape)
```

```python
# Import train_test_split function
from sklearn.model_selection import train_test_split

# Split dataset into training set and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=109)
```

```python
from sklearn.neighbors import KNeighborsClassifier

from  sklearn import metrics

k_range = range(1,26)
scores = {}
scores_list = []

for k in k_range:
    knn = KNeighborsClassifier(n_neighbors = k)
    knn.fit(X_train,y_train)
    y_pred=knn.predict(X_test)
    scores[k] = metrics.accuracy_score(y_test,y_pred)
    scores_list.append(metrics.accuracy_score(y_test,y_pred))
```

```python
import matplotlib.pyplot as plt
plt.plot(k_range,scores_list)
plt.xlabel('K')
plt.ylabel('Accuracy')
```

```python
knn = KNeighborsClassifier(n_neighbors=7)
knn.fit(X_train,y_train)
```

```python
y_predict = knn.predict(X_test)
```

```python
pd.DataFrame({'Prediction':y_predict, 'Acutal':y_test})
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_predict, labels = ['disconnected','connected'])
```

```python
# import dependencies
import pandas as pd
```

```python
#Read csvs:
disconnected_df = pd.read_csv('disconnected_final.csv')
connected_df = pd.read_csv('connected_final.csv')
```

```python
#Get matches 
address_match = 0
city_match = 0
zipcode_match = 0
for row in disconnected_df.itertuples():
    if row.Address == row.OFFICE_ADDRESS_LINE_2:
        address_match += 1
    if row.City == row.OFFICE_ADDRESS_CITY:
        city_match += 1
    if row.Zipcode == row.OFFICE_ADDRESS_ZIP:
        zipcode_match += 1
print (f'{address_match} addresses match  {city_match} cities match  {zipcode_match} zipcodes match')
```

```python
#Get matches 2
address_match = 0
city_match = 0
zipcode_match = 0
for row in connected_df.itertuples():
    if row.Address == row.OFFICE_ADDRESS_LINE_2:
        address_match += 1
    if row.City == row.OFFICE_ADDRESS_CITY:
        city_match += 1
    if row.Zipcode == row.OFFICE_ADDRESS_ZIP:
        zipcode_match += 1
print (f'{address_match} addresses match  {city_match} cities match  {zipcode_match} zipcodes match')
```

```python
#def med words

med_words = ['cancer', 'cardiology', 'neurology', 'family care', 'pulmonary', 'anesthesia', 'orthope', 'urgent care', 'allergy', 'kidney', 'surgery', 'hosp', 'mri', 'throat', 'dentist', 'med', 'clinic', 'health', 'gastroenter','anesthesiologist','patient','physician','surgeon','doctor','hospital', 'md', 'medical']
```

```python
#How many names contain med words?

healthcount = 0
false_neg=[]
for row in connected_df.itertuples():
    healthy = False
    for word in med_words:
        if word in row.Name.lower():
            healthy = True
    if healthy == True:    
        healthcount += 1
    else:
        false_neg.append(row.Name)

print(f'TP:{healthcount}  FN:{100 - healthcount} ')
```

```python
#Checking to see if algoorithm missed any medical words
false_neg
```

```python
#Medical word count for disconnnected numbers

healthcount = 0
false_pos=[]
neg =[]
for row in disconnected_df.itertuples():
    healthy = False
    for word in med_words:
        if word in str(row.Name).lower():
            healthy = True
    if healthy == True:    
        healthcount += 1
        false_pos.append(row.Name)
    else:
        neg.append(row.Name)

print(f'FP:{healthcount}  TN:{100 - healthcount} ')
```

```python
mailcount = 0
connectcount = 0
portcount = 0
novalidcount = 0
unknowncount = 0
disconnectedcount = 0
voipcount = 0
wirelesscount = 0

for row in connected_df.itertuples():
        if 'IsMailable' in str(row.Notes):
            mailcount += 1
        if 'IsConnected' in str(row.Notes):
            connectcount += 1
        if 'IsPorted' in str(row.Notes):
            portcount += 1
        if 'NotValid' in str(row.Notes):
            novalidcount += 1
        if 'IsUnknownContact' in str(row.Notes):
            unknowncount += 1
        if 'IsPossibleDisconnected' in str(row.Notes):
            disconnectedcount += 1
        if 'IsPossiblePortableVOIP' in str(row.Notes):
            voipcount += 1
        if 'IsWireless' in str(row.Notes):
            wirelesscount += 1
            
print(f'{mailcount} {connectcount} {portcount} {novalidcount} {unknowncount} {disconnectedcount} {voipcount} {wirelesscount}')
```

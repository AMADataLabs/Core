```python
all_response_json_list = []
all_phone_df_list = []
```

```python
test_dataset.shape
```

```python
test_dataset.head()
```

```python
def query_api_on_data(query_dataframe):
    for i, row in query_dataframe.iterrows():
        me_number = row.ME_NO
        print('ME_NO:', me_number)

        params = get_residential_params(row)
        response = requests.get(residential_url, params=params)

        ### Logging ###
        me_no_response = {
            'ME_NO': me_number,
            'response_json': response.json()
        }
        all_response_json_list.append(me_no_response)
        print(response.json()['Phones'])

        ### Parsing ###
        response_phone_data = get_phone_numbers_from_response(response)
        response_phone_data['ME_NO'] = me_number

        all_phone_df_list.append(response_phone_data)

        print('#'*80)
        print()

        sleep(1)
```

```python
query_api_on_data(test_dataset)
```

```python
combined_data = pd.concat(all_phone_df_list, ignore_index=True)
combined_data
```

```python
combined_data['AccurateAppend_Phone'] = combined_data['AreaCode'] + combined_data['PhoneNumber']
```

### Accuracy

```python
merged = test_dataset[['ME_NO', 'OFFICE_PHONE']].merge(
    combined_data[['ME_NO', 'AccurateAppend_Phone']],
    on='ME_NO',
    how='left'
)
merged['match'] = merged['OFFICE_PHONE'] == merged['AccurateAppend_Phone']
merged
```

```python
merged['match_partial'] = [str(o)[:6] == str(a)[:6] for o, a in zip(merged['OFFICE_PHONE'], merged['AccurateAppend_Phone'])]
```

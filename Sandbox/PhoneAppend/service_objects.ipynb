```python
INDEX= 0
dict_list_3 =[]
all_results_2 =[]
for row in new_sample[0:].itertuples():
    print(INDEX)
    print(f'{row.FIRST_NAME} {row.LAST_NAME}')
    parameters = {
    'FirstName': row.FIRST_NAME,
    'LastName':row.LAST_NAME,
    'Address': row.POLO_MAILING_LINE_2,
    'City': row.POLO_CITY,
    'State': row.POLO_STATE,
    'PostalCode': row.POLO_ZIP,
    'LicenseKey': key}

    response =  requests.get(url, params=parameters)
    results = xmltodict.parse(response.content)
    try:
        new_dict={
            'ME': row.ME,
            'phone': results['PhoneInfoResponse']['PhoneInfo']['Phone'],
            'name': results['PhoneInfoResponse']['PhoneInfo']['Name'],
            'address': results['PhoneInfoResponse']['PhoneInfo']['Address'],
            'city': results['PhoneInfoResponse']['PhoneInfo']['City'],
            'state': results['PhoneInfoResponse']['PhoneInfo']['State'],
            'postal_code': results['PhoneInfoResponse']['PhoneInfo']['PostalCode'],
            'residential':results['PhoneInfoResponse']['PhoneInfo']['IsResidential'],
            'certainty':results['PhoneInfoResponse']['PhoneInfo']['Certainty'],
            'linetype':results['PhoneInfoResponse']['PhoneInfo']['LineType']
        }
        dict_list_3.append(new_dict)
        print('FOUND!')
        print(f"{new_dict['name']} : {new_dict['phone']}")
    except:
        print(results['PhoneInfoResponse']['Error'])
    all_results_2.append(results)
    INDEX += 1
```

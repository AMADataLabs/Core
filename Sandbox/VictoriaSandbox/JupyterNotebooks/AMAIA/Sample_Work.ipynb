```python
import pandas as pd
```

```python
df = pd.read_csv('Insurance_Sample_2.csv')
```

```python
df.columns
```

```python
pure = df[(df.Match==False)&(df.Phone_Number_mar!='None')]
```

```python
df[['CITY','CITY_PPMA','CITY_POLO']]
```

```python
count = 0
match = []
for row in df.itertuples():
    if row.CITY != row.CITY_PPMA and row.CITY!= row.CITY_POLO:
        count += 1
        match.append(False)
    else:
        match.append(True)
```

```python
df['CITY_MATCH']=match
```

```python
impure = df[(df.Match==True)&(df.Phone_Number_mar!='None')&(df.CITY_MATCH==False)]
```

```python
pure.drop_duplicates('Phone_Number_mar')
```

```python
impure_2 = impure.drop_duplicates('Phone_Number_mar').head(212)
```

```python
pd.concat([pure, impure_2]).drop_duplicates('Phone_Number_mar', keep='first').to_csv('True_Sample.csv',index=False)
```

```python
import pandas as pd
from fuzzywuzzy import fuzz
```

```python
ws = pd.read_csv('wslive_with_results_2019-01-04_to_2020-01-24.csv')
```

```python
dhc = pd.read_csv('DHC_March_Delta.csv')
```

```python
DHC = dhc[dhc.ME_mar!='None']
```

```python
mes = list(DHC['ME_mar'].astype(float).astype('int64'))
DHC['PHYSICIAN_ME_NUMBER'] = mes
```

```python
ws.groupby('addr_status').count()
```

```python
TESTY = pd.merge(ws, DHC, on = 'PHYSICIAN_ME_NUMBER')
```

```python
TESTY[['Address_mar', 'Address1_mar','OFFICE_ADDRESS_LINE_2','Phone_Number_mar','OFFICE_TELEPHONE']]
```

```python
TESTY = TESTY.fillna('None')
```

```python
TESTY['OFFICE_TELEPHONE'].astype(str)
```

```python
phone_match = []
add_match = []
for row in TESTY.itertuples():
    if str(row.Phone_Number_mar).replace('.','') == str(row.OFFICE_TELEPHONE).replace('.0',''):
        phone_match.append(True)
        print('Phone match')
    else:
        phone_match.append(False)
    if fuzz.ratio(row.Address_mar.upper(),row.OFFICE_ADDRESS_LINE_2)>70:
        add_match.append(True)
        print('Address match')
    else:
        add_match.append(False)
```

```python
TESTY['DHC_PHONE_MATCH']=phone_match
TESTY['DHC_ADDRESS_MATCH']=add_match
```

```python
TESTY[(TESTY.ws_year==2020)&(TESTY.DHC_ADDRESS_MATCH==False)&(TESTY.addr_status.isin(['Confirmed','Updated']))][['City_mar','OFFICE_ADDRESS_LINE_2']]
```

```python
TESTY[(TESTY.ws_year==2020)&(TESTY.Phone_Update==True)&(TESTY.DHC_PHONE_MATCH==False)&(TESTY.phone_status.isin(['Confirmed','Updated']))][['OFFICE_TELEPHONE','Phone_Number_mar','Phone_Number_feb']]
```

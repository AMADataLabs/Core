```python
mf_Samp = pd.read_excel('Masterfile_Random_Sample.xlsx')
```

```python
CODES = mf_Samp.drop_duplicates('PE_CD')[['DESCRIPTION','PE_CD']]
```

```python
mf_Samp.columns
```

```python
test.columns
```

```python
TEST = test.rename(columns = {'ADDR_2':'POLO_MAILING_LINE_1','ADDR_1':'POLO_MAILING_LINE_2','CITY_y':'POLO_CITY','STATE_CD':'POLO_STATE','ZIP_y':'POLO_ZIP','Phone_Number_mar':'TELEPHONE_NUMBER'})
```

```python
WORKIN = pd.merge(TEST, CODES,on='PE_CD')
```

```python
WORKIN = WORKIN.fillna('None')
```

```python
new_list =[]
for thing in list(WORKIN['TELEPHONE_NUMBER']):
    new_list.append(thing.replace('.',''))
WORKIN['TELEPHONE_NUMBER'] = new_list
```

```python
WORKIN.to_csv('2020_03_24_Insurance_DHC_Sample.csv',index=False)
```

```python
WORKIN = WORKIN.astype(str)
```

```python
nums =[]
for num in list(WORKIN['ME'].astype(str)):
    num = num.replace('.0','')
    if len(num)==10:
        num = '0'+num
    elif len(num)==9:
        num = '00'+num
    elif len(num)==8:
        num = '000'+num
    nums.append(num)
WORKIN['ME']=nums
```

```python
WORKIN.to_excel('2020_03_24_Insurance_DHC_Sample.xlsx')
```

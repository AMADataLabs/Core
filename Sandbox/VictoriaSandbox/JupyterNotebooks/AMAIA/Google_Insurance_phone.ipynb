```python
for num in phones[10:]:
    print("")
    print('----------------------')
    phone = '+1' + num
    parameters = {'fields': fields, 'inputtype':inputtype, 'key':key,'input': phone}
    response =  requests.get(base_url, params=parameters)
    if response.status_code == 200:
        results = response.json()
        all_results.append(results)
        if len(results['candidates'])>0:
            for candidate in results['candidates']:
                new_dict ={}
                new_dict['TELEPHONE_NUMBER'] = num
                new_dict['Name'] = candidate["name"]
                new_dict['Address'] = candidate['formatted_address']
                new_dict['Types'] = candidate['types']
                list_of_dicts.append(new_dict)
                index = results['candidates'].index(candidate)
                print(f'{index + 1}. {candidate["name"]}')
        else:   
            print(f'{phone} not found')
            formatted_address = 'None'
            name = 'None'
            types = 'None'
            new_dict ={}
            new_dict['TELEPHONE_NUMBER'] = num
            new_dict['Name'] = name
            new_dict['Address'] = formatted_address
            new_dict['Types'] = types
            list_of_dicts.append(new_dict)
    else:
        print(f'{phone} not found')
        new_dict ={}
        new_dict['TELEPHONE_NUMBER'] = num
        formatted_address = 'None'
        name = 'None'
        types = 'None'
        new_dict['Name'] = name
        new_dict['Address'] = formatted_address
        new_dict['Types'] = types
        list_of_dicts.append(new_dict)
```

```python
BIG_MERGE = pd.merge(google_df, df, on='TELEPHONE_NUMBER')
```

```python
BIG_MERGE = BIG_MERGE.rename(columns={'Physician Name_mar':'Physician_Name_Mar'})
```

```python
add_match =0
city_match =0
state_match = 0
zip_match = 0
name_match = 0
for row in BIG_MERGE.itertuples():
    if fuzz.ratio(row.ADDR_1, row.Address.split(',')[0])>70:
        add_match += 1
    if row.CITY in row.Address:
        city_match+= 1
    if row.STATE_CD in row.Address:
        state_match += 1
    if row.ZIP in row.Address:
        zip_match += 1
    if fuzz.ratio(row.Physician_Name_Mar, row.Name)>70:
        name_match += 1
```

```python
me_list =[]
name_match = 0
g_match_list =[]
for row in BIG_MERGE.itertuples():
    g_match = False
    if row.ME not in me_list:
        if fuzz.ratio(row.Physician_Name_Mar, row.Name)>70:
            print(f'MATCH: {row.Physician_Name_Mar} matches {row.Name}')
            name_match += 1
            me_list.append(row.ME)
            g_match = True
    g_match_list.append(g_match)
```

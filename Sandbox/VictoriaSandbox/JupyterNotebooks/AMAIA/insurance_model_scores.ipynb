```python
import pandas as pd
from fuzzywuzzy import fuzz
from fuzzywuzzy import process
import datetime as dt
```

```python
ranked = pd.read_csv('Polo_Addr_Rank_Scored_PPD_DPC_Pop.csv')
dhc_insurance = pd.read_csv('dhc_insurance.csv')slyth
```

```python
ALL_DATA = pd.merge(dhc_insurance, ranked, left_on = 'ME', right_on = 'PPD_ME')
```

```python
ppd = pd.read_csv('ppd_data_20200222.csv')
PPD  = ppd[['ME','MAILING_LINE_1','MAILING_LINE_2','CITY','STATE','ZIP']]
```

```python
all_ = pd.merge(ALL_DATA, PPD, on = 'ME')
```

```python
all_ = all_.fillna('None')
```

```python
ppma_amaia_match = []
polo_dhc_match = []
polo_amaia_match = []
dhc_amaia_match = []
for row in all_.itertuples():
    if type(row.ADDR_1) == float:
        print(row.ADDR_1)
    if fuzz.ratio(row.ADDR_1.upper(), row.PPD_POLO_MAILING_LINE_2) >75:
        polo_amaia_match.append(True)
    else:
        polo_amaia_match.append(False)
    if fuzz.ratio(row.Address.upper(),row.PPD_POLO_MAILING_LINE_2) > 75:
        polo_dhc_match.append(True)
    else:
        polo_dhc_match.append(False)
    if fuzz.ratio(row.ADDR_1, row.Address) >75:
        dhc_amaia_match.append(True)
    else:
        dhc_amaia_match.append(False)
    if fuzz.ratio(row.ADDR_1.upper(), row.MAILING_LINE_2) >70:
        ppma_amaia_match.append(True)
    else:
        ppma_amaia_match.append(False)
```

```python
all_['PPMA_AMAIA'] = ppma_amaia_match
all_['POLO_AMAIA'] = polo_amaia_match
all_['DHC_AMAIA'] = dhc_amaia_match
all_['POLO_DHC'] = polo_dhc_match
```

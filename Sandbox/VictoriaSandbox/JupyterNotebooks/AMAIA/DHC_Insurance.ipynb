```python
import pandas as pd
from fuzzywuzzy import fuzz
from fuzzywuzzy import process
import datetime as dt
```

```python
data = pd.read_csv('dhc_insurance.csv')
```

```python
data = data.rename(columns={'Zip Code':'ZipCode'})
data = data.fillna('None')
```

```python
LIST_ME =[]
COUNT = 0
for row in data.itertuples():
    if fuzz.ratio(row.ADDR_1, row.Address) > 71:
        COUNT+= 1
        LIST_ME.append(row.ME)
COUNT/len(data)*100
```

```python
cities =[]
for ci in list(data.CITY):
    ci = ci.replace('  ','')
    cities.append(ci)
data.CITY = cities
```

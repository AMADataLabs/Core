```python
MD_COUNT = 0
DO_COUNT = 0
not_important_count = 0
for row in dhc_phys.itertuples():
    relevant = False
    if 'MD' in str(row.CREDENTIAL):
        MD_COUNT += 1
        relevant = True
    if 'DO' in str(row.CREDENTIAL):
        DO_COUNT += 1
        relevant = True
    if relevant == False:
        not_important_count += 1
print(f'There are {MD_COUNT} MDs, {DO_COUNT} DOs in the DHC data.\n{not_important_count} are neither DOs or MDs')
```

```python
def count_docs(df):
    MD_COUNT = 0
    DO_COUNT = 0
    not_important_count = 0
    for row in df.itertuples():
        relevant = False
        if 'MD' in str(row.CREDENTIAL):
            MD_COUNT += 1
            relevant = True
        if 'DO' in str(row.CREDENTIAL):
            DO_COUNT += 1
            relevant = True
        if relevant == False:
            not_important_count += 1
    print(f'There are {MD_COUNT} MDs and {DO_COUNT} DOs in the DHC data.\n{not_important_count} are neither DOs or MDs')
```

```python
bad_list = []
polo_match = 0
ppma_match = 0
match_count = 0
polo_total = dhc_ppd['POLO_MAILING_LINE_2'].count()
total = len(dhc_ppd)
for row in dhc_ppd.itertuples():
    match = False
    polo = str(row.POLO_CITY).upper()
    ppma = str(row.CITY).upper()
    loc_list = [str(row.HQ_CITY).upper(), str(row.HQ_CITY_1).upper(), str(row.HQ_CITY_2).upper(), str(row.HQ_CITY_3).upper(), str(row.HQ_CITY_4).upper(), str(row.HQ_CITY_5).upper(), str(row.HQ_CITY_6).upper(), str(row.HQ_CITY_7).upper()]
    if polo in loc_list:
        polo_match += 1
        match =True
    if ppma in loc_list:
        ppma_match += 1
        match=True
    if match == False:
        match_count += 1
        bad_list.append(row.ME)
print(f'{polo_match/polo_total*100} polo cities match')
print(f'{ppma_match/total*100} ppma cities match')
print(f'{match_count/total*100} matched neither')
```

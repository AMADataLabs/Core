```python
import pandas as pd
```

```python
all_connected = pd.read_csv("GetPhoneInfo/connected_final.csv")
```

```python
ekata_connected = pd.read_csv("Ekata/ekata_connected.csv")
everyone_connected = pd.read_csv("Telo/everyone_numbers_connected.csv")
```

```python
base_numbers = pd.merge(ekata_connected, everyone_connected, on = 'OFFICE_TELEPHONE', how = 'outer')
```

```python
base_numbers
```

```python
x = pd.merge(all_connected, ekata_connected, on = 'OFFICE_TELEPHONE', how = 'outer')
```

```python
potential_new_numbers = x[x['Error']!='None']
```

```python
new_numbers= potential_new_numbers.sample(n=10)
```

```python
new_numbers.to_csv("new_connected_numbers.csv")
```

```python
ekata_disconnected = pd.read_csv("Ekata/ekata_disconnected.csv")
everyone_disconnected = pd.read_csv("Telo/everyone_numbers_disconnected.csv")
```

```python
all_disconnected = pd.read_csv("GetPhoneInfo/disconnected_final.csv")
```

```python
y = pd.merge(all_disconnected, ekata_disconnected, on = 'OFFICE_TELEPHONE', how = 'outer')
potential_new_numbers2 = y[y['Error']!='None']
new_numbers2= potential_new_numbers2.sample(n=10)
new_numbers2.to_csv("new_disconnected_numbers.csv")
```

```python
everyone_wrong = pd.read_csv('Telo/everyone_numbers_wrong.csv')
all_wrong = pd.read_csv('Telo/wrong_CNAM.csv')
z = pd.merge(all_wrong, everyone_wrong, on = 'OFFICE_TELEPHONE', how = 'outer')
```

```python
potential_numbers3 = z[z['STATUS']!='WRONG NUMBER']
new_numbers3= potential_numbers3.sample(n=40)
new_numbers3.to_csv("new_wrong_numbers.csv")
```

```python
def analyze_numbers(dataframe):
    
    address_match = 0
    city_match = 0
    zipcode_match = 0
    healthcount = 0
    mailcount = 0
    connectcount = 0
    portcount = 0
    unknowncount = 0
    disconnectedcount = 0
    voipcount = 0
    wirelesscount = 0
    no_address_count = 0
    no_date_count = 0
    business_count = 0
    residential_count = 0
    low_count = 0
    high_count = 0
    med_count = 0
    first_name_match = 0
    last_name_match = 0 
    workplace_match = 0
    state_match = 0
    no_name = 0
    tiny_count =0 


    for row in dataframe.itertuples():
        gpi_add = g.translate_addr(str(row.ContactAddressOut))
        ws_add = g.translate_addr(str(row.OFFICE_ADDRESS_LINE_2))
        if gpi_add != 'NONE':
            if gpi_add in ws_add or ws_add in gpi_add:
                address_match += 1
        if str(row.OFFICE_ADDRESS_CITY) in str(row.ContactCityOut):
            city_match += 1
        if g.clean_zipcode(row.OFFICE_ADDRESS_ZIP) == g.clean_zipcode(row.ContactZipOut):
            zipcode_match += 1

        if row.ContactStateOut == row.OFFICE_ADDRESS_STATE:
            state_match += 1

        if str(row.PHYSICIAN_LAST_NAME).lower() in str(row.ContactNameOut).lower():
            last_name_match +=1
        if str(row.PHYSICIAN_LAST_NAME).lower() in str(row.ContactNameOut).lower() or str(row.OFFICE_ADDRESS_LINE_1).lower() in str(row.ContactNameOut).lower():
            tiny_count += 1
        if str(row.PHYSICIAN_FIRST_NAME).lower() in str(row.ContactNameOut).lower():
            first_name_match +=1

        if str(row.OFFICE_ADDRESS_LINE_1).lower() in str(row.ContactNameOut).lower():
            workplace_match += 1
#         if str(row.OFFICE_ADDRESS_LINE_1).lower() not in str(row.ContactNameOut).lower():
#             print (f'{row.OFFICE_ADDRESS_LINE_1} is not {row.ContactNameOut}')
        if is_healthy(row.ContactNameOut)==True: 
            healthcount += 1
                
            

        if 'IsMailable' in str(row.NoteDescriptions):
            mailcount += 1
        if 'IsConnected' in str(row.NoteDescriptions):
            connectcount += 1
        if 'IsPorted' in str(row.NoteDescriptions):
            portcount += 1
        if 'IsUnknownContact' in str(row.NoteDescriptions):
            unknowncount += 1
        if 'IsPossibleDisconnected' in str(row.NoteDescriptions):
            disconnectedcount += 1
#             if is_healthy(row.ContactNameOut)==True:  
#                 print(row.ContactNameOut)
        if 'IsPossiblePortableVOIP' in str(row.NoteDescriptions):
            voipcount += 1
        if 'IsWireless' in str(row.NoteDescriptions):
            wirelesscount += 1

        if row.ContactAddressOut == 'None':
            no_address_count += 1
        if row.DateOfPorting == 'None':
            no_date_count += 1
        if row.ContactPhoneType == 'BUSINESS':
            business_count += 1
        if row.ContactPhoneType == 'RESIDENTIAL':
            residential_count += 1
        if row.ContactQualityScore == 'LOW':
            low_count += 1
        if row.ContactQualityScore == 'HIGH':
            high_count += 1
        if row.ContactQualityScore == 'MED':
            med_count += 1
        if row.ContactNameOut =='None':
            no_name += 1

    new_dict = {
        'Address Matches': address_match,
        'City Matches': city_match,
        'ZipCode Matches': zipcode_match,
        'No Address': no_address_count,
        'No Date': no_date_count,
        'Relevant Name': healthcount,
        'Business Phone': business_count,
        'Residential Phone': residential_count,
        'Low Quality': low_count,
        'Medium Quality': med_count,
        'High Quality': high_count,
        'Mailable': mailcount,
        'Connected': connectcount,
        'Ported': portcount,
        'Unknown Contact': unknowncount,
        'Possibly Disconnected': disconnectedcount,
        'Possibly Portable VOIP': voipcount,
        'Wireless': wirelesscount,
        'First Name Match':first_name_match,
        'Last Name Match':last_name_match,
        'Workplace Match': workplace_match,
        'State Match': state_match, 
        'No Name': no_name,
        'New Count': tiny_count
#         'disconnected':rpv_disconnected
        }
        
    return(new_dict)
```

```python
def check_variables(row,variable_pair):
    
    address_no_match = True
    zipcode_no_match = False
    city_no_match = False
    unknowncount = False
    disconnectedcount = False
    no_first_name_match = False
    no_last_name_match = False
    no_workplace_match = False
    state_no_match = False

    gpi_add = g.translate_addr(str(row.ContactAddressOut))
    ws_add = g.translate_addr(str(row.OFFICE_ADDRESS_LINE_2))
    if gpi_add != 'NONE':
        if gpi_add in ws_add or ws_add in gpi_add:
            address_no_match = False
    if g.clean_zipcode(row.OFFICE_ADDRESS_ZIP) != g.clean_zipcode(row.ContactZipOut):
        zipcode_no_match = True
    if 'IsUnknownContact' in str(row.NoteDescriptions):
        unknowncount = True
    if 'IsPossibleDisconnected' in str(row.NoteDescriptions):
        disconnectedcount = True
        
    if str(row.OFFICE_ADDRESS_CITY) not in str(row.ContactCityOut):
        city_no_match = True

    if row.ContactStateOut != row.OFFICE_ADDRESS_STATE:
        state_no_match = True

    if str(row.PHYSICIAN_LAST_NAME).lower() not in str(row.ContactNameOut).lower():
        no_last_name_match = True
    if str(row.PHYSICIAN_FIRST_NAME).lower() not in str(row.ContactNameOut).lower():
        no_first_name_match = True

    if str(row.OFFICE_ADDRESS_LINE_1).lower() not in str(row.ContactNameOut).lower():
        no_workplace_match = True
        
    new_dict = {
        'Address Mismatches': address_no_match,
        'ZipCode Mismatches': zipcode_no_match,
        'Unknown Contact': unknowncount,
        'Possibly Disconnected': disconnectedcount,
        'City Mismatches': city_no_match,
        'First Name Mismatch':no_first_name_match,
        'Last Name Mismatch':no_last_name_match,
        'Workplace Mismatch': no_workplace_match,
        'State Mismatch': state_no_match, 
        }
    conclusion = False
    if new_dict[variable_pair[0]]==True or new_dict[variable_pair[1]]==True:
        if new_dict[variable_pair[2]]==True and new_dict[variable_pair[3]]==True:
                conclusion=True
    return(conclusion)
```

```python
def try_variables(pair_list, df):
    list_of_dicts =[]
    for pair in pair_list:
        new_dict={}
        new_dict['Pair']=pair
        count = 0 
        for row in df.itertuples():
#             if row.status=='disconnected' or row.status=='disconnected-70':
            if check_variables(row, pair)==True:
                count += 1
        new_dict['Count']=count
        list_of_dicts.append(new_dict)
    return(list_of_dicts)
```

```python
def variable_breakdown(dataframe):
    
    LIST_OF_DICTS = []
    
    
    for row in dataframe.itertuples():

        address_no_match = 1
        zipcode_no_match = 0
        city_no_match = 0
        unknowncount = 0
        disconnectedcount = 0
        no_first_name_match = 0
        no_last_name_match = 0
        no_workplace_match = 0
        state_no_match = 0
        
        low_quality = 0
        no_date  = 0
        not_relevant = 0
        not_high_quality = 0
        not_mailable = 0 
        residential = 0
        no_name = 0 
        no_address = 0 
        not_connected = 0
        not_ported = 0 
        inf = 0 
        

        gpi_add = g.translate_addr(str(row.ContactAddressOut))
        ws_add = g.translate_addr(str(row.OFFICE_ADDRESS_LINE_2))
        if gpi_add != 'NONE':
            if gpi_add in ws_add or ws_add in gpi_add:
                address_no_match = 0
        if g.clean_zipcode(row.OFFICE_ADDRESS_ZIP) != g.clean_zipcode(row.ContactZipOut):
            zipcode_no_match = 1
        if 'IsUnknownContact' in str(row.NoteDescriptions):
            unknowncount = 1
        if 'IsPossibleDisconnected' in str(row.NoteDescriptions):
            disconnectedcount = 1

        if str(row.OFFICE_ADDRESS_CITY) not in str(row.ContactCityOut):
            city_no_match = 1

        if row.ContactStateOut != row.OFFICE_ADDRESS_STATE:
            state_no_match = 1

        if str(row.PHYSICIAN_LAST_NAME).lower() not in str(row.ContactNameOut).lower():
            no_last_name_match = 1
        if str(row.PHYSICIAN_FIRST_NAME).lower() not in str(row.ContactNameOut).lower():
            no_first_name_match = 1

        if str(row.OFFICE_ADDRESS_LINE_1).lower() not in str(row.ContactNameOut).lower():
            no_workplace_match = 1
            
        if g.is_healthy(row.ContactNameOut)==False: 
            not_relevant = 1
                
            

        if 'IsMailable' not in str(row.NoteDescriptions):
            not_mailable = 1
        if 'IsConnected' not in str(row.NoteDescriptions):
            not_connected = 1
        if 'IsPorted' not in str(row.NoteDescriptions):
            not_ported = 1

        if row.ContactAddressOut == 'None':
            no_address = 1
        if row.DateOfPorting == 'None':
            no_date = 1

        if row.ContactPhoneType == 'RESIDENTIAL':
            residential = 1
        if row.ContactQualityScore == 'LOW':
            low_quality = 1
        if row.ContactQualityScore != 'HIGH':
            not_high_quality = 1

        if row.ContactNameOut =='None':
            no_name = 1
                
        if 'INF' in str(row.NoteDescriptions):
            inf = 1


        new_dict = {
            'OFFICE_TELEPHONE': row.OFFICE_TELEPHONE,
            'AddressMismatch': address_no_match,
            'ZipCodeMismatch': zipcode_no_match,
            'UnknownContact': unknowncount,
            'PossiblyDisconnected': disconnectedcount,
            'CityMismatch': city_no_match,
            'FirstNameMismatch':no_first_name_match,
            'LastNameMismatch':no_last_name_match,
            'WorkplaceMismatch': no_workplace_match,
            'StateMismatch': state_no_match, 
            'NoAddress': no_address,
            'NoDate': no_date,
            'IrrelevantName': not_relevant,
            'ResidentialPhone': residential,
            'LowQuality': low_quality,
            'NotHighQuality': not_high_quality,
            'NotMailable': not_mailable,
            'NotConnected': not_connected,
            'NotPorted': not_ported,
            'NoName': no_name,
            'INF': inf,
            'Label': row.COMMENTS,
            'Score': row.pred_probability,
            'Class': row.pred_class,
            'ME': row.ppd_me,
            'Description': row.pe_description
            }

        LIST_OF_DICTS.append(new_dict)

    return(LIST_OF_DICTS)
```

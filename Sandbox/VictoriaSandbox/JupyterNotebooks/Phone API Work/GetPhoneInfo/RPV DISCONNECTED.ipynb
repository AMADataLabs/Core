```python
import pandas as pd
```

```python
RPV = pd.read_csv('../RPV_results_all.csv')
```

```python
X = RPV[(RPV.status =='disconnected') | (RPV.status =='disconnected-70')]
```

```python
disconnected = X[X.DISCONNECTED == True]
```

```python
X.shape
```

```python
connected = X[X.DISCONNECTED == False]
```

```python
nums = pd.read_csv('../../Numbers.csv')
```

```python
ALREADY_DONE = pd.merge(X, nums, on='OFFICE_TELEPHONE')
```

```python
ALREADY_DONE.groupby('DISCONNECTED').count()
```

```python
disconnect_df = pd.merge(disconnected, ALREADY_DONE, how='outer', indicator=True)
disconnect_df = disconnect_df[disconnect_df._merge == 'left_only']
disconnect_df = disconnect_df[['OFFICE_TELEPHONE','COMMENTS','DISCONNECTED','status','rpv_dc']]
```

```python
connect_df = pd.merge(connected, ALREADY_DONE, how='outer', indicator=True)
connect_df = connect_df[connect_df._merge == 'left_only']
connect_df = connect_df[['OFFICE_TELEPHONE','COMMENTS','DISCONNECTED','status','rpv_dc']]
```

```python
connected_sample = connect_df.sample(141)
```

```python
disconnected_sample = disconnect_df.sample(97)
```

```python
wslive = pd.read_csv('../../wslive_with_results_2019-08-02_to_2019-08-23.csv')
```

```python
WSLIVE = wslive[['PHYSICIAN_ME_NUMBER','PHYSICIAN_FIRST_NAME','PHYSICIAN_MIDDLE_NAME','PHYSICIAN_LAST_NAME','OFFICE_ADDRESS_LINE_1',
       'OFFICE_ADDRESS_LINE_2','OFFICE_ADDRESS_CITY','OFFICE_ADDRESS_STATE','OFFICE_ADDRESS_ZIP','OFFICE_TELEPHONE']]
```

```python
wslive_no_duplicates = wslive.drop_duplicates('OFFICE_TELEPHONE')
```

```python
connected_new = pd.merge(connected_sample, wslive_no_duplicates, on= 'OFFICE_TELEPHONE')
```

```python
disconnected_new = pd.merge(disconnected_sample, wslive_no_duplicates, on= 'OFFICE_TELEPHONE')
```

```python
connected_new.to_csv('connected_rpv_getphone.csv',index=False)
disconnected_new.to_csv('disconnected_rpv_getphone.csv',index=False)
```

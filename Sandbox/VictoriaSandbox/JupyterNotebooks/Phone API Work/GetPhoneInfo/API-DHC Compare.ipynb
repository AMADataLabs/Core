```python
count =0
other_count = 0
for row in w_dhc.itertuples():
    dhc_add_match =False

    if str(row.Phone_Number.replace('.','')) == str(row.OFFICE_TELEPHONE):
        other_count += 1
        add_dhc = row.Address_DHC.upper()
        add_gfi = row.Address_MF
        if add_dhc in add_gfi or add_dhc in add_gfi:
            dhc_add_match = True
        elif fuzz.ratio(add_dhc,add_gfi)>70:
            dhc_add_match =True
        if dhc_add_match == True:
            count += 1
```

```python
count/other_count
```

```python
count =0
other_count = 0
for row in w_dhc.itertuples():
    gfi_add_match =False

    if str(row.Phone_Number.replace('.','')) == str(row.OFFICE_TELEPHONE):
        if row.Google_Address != 'None':
            other_count += 1
            add_dhc = row.Address_DHC.upper()

            adds = row.Google_Addresses.replace('"',"'").replace("['","").replace("']","")
            all_adds = adds.split("', '")
            for add in all_adds:
                add_gfi = add.split(', ')[0].upper()
                if add_dhc in add_gfi or add_dhc in add_gfi:
                    gfi_add_match = True
                    break
                elif fuzz.ratio(add_dhc,add_gfi)>87:
                    gfi_add_match =True
                    break

            if gfi_add_match == True:
                count += 1
```

```python
count = 0 
other_count = 0
for row in all_.itertuples():
    add_match = False
    other_count += 1
    add_mf = row.OFFICE_ADDRESS_LINE_2.upper()
    add_gfi = row.Address.upper()
    if add_mf in add_gfi or add_mf in add_gfi:
        add_match = True
    elif fuzz.ratio(add_mf,add_gfi)>70:
        add_match =True
    if add_match == True:
        count += 1
```

```python
dhc = pd.read_csv('DHC_Data/DHC_Data.csv')
```

```python
append = pd.read_csv('Accurate_Append.csv')
```

```python
MERGE = pd.merge(dhc, append, on='ME')
```

```python
MERGE = MERGE.rename(columns={'Phone Number':'PhoneNumber1','Physician Name':'PhysicianName'})
MERGE=MERGE.fillna('None')
```

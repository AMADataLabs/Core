```python
import pandas as pd
```

```python
rpv = pd.read_csv('../RPV_archive.csv')
getphone = pd.read_csv('../GetPhoneInfo/GetPhoneInfoAPIResults.csv')
```

```python
MERGED = pd.merge(rpv, getphone, left_on = 'phone', right_on = 'OFFICE_TELEPHONE')
```

```python
def match_carriers(df):
    total=df.shape[0]
    count = 0 
    other_count = 0
    connected = 0
    disconnected = 0
    dis_total = df[df['WSLive_Status']=='disconnected'].shape[0]
    con_total = df[df['WSLive_Status']=='connected'].shape[0]
    for row in df.itertuples():
        RPV_list = str(row.carrier).lower().split(' ')
        GetPhone_list = str(row.Provider).lower().split(' ')
        mismatched = True
        
        for word in RPV_list:
            if word in GetPhone_list:
                mismatched = False
#         if str(row.carrier).lower() not in str(row.Provider).lower() and str(row.carrier).lower() not in str(row.Provider).lower():
#             count += 1
        if mismatched==True:
            other_count += 1
            print(f'\nRPV:{row.status} Actual:{row.WSLive_Status}\n--------\n{row.carrier} is not {row.Provider}')
            if row.WSLive_Status =='connected':
                connected+=1
            else:
                disconnected +=1
#     print(count/total)
    print(other_count/total*100)
    print(f'{connected/con_total*100} connected numbers had mismatching carriers')
    print(f'{disconnected/dis_total*100} disconnected numbers had mismatching carriers')
```

```python
wrong = pd.read_csv("../new_wrong_numbers_all.csv")
```

```python
def twilio_lookup(number):
    new_dict["OFFICE_TELEPHONE"] = number
    url = f'https://lookups.twilio.com/v1/PhoneNumbers/+1{number}?Type=caller-name'
    response =  requests.get(url, auth=(ACCOUNT_SID, your_auth_token))
    results = response.json()
    try:
        name = results['caller_name']['caller_name']
    except:
        name = 'No Name'
    try:
        caller_type = results['caller_name']['caller_type']
    except:
        caller_type = "No caller type"
    new_dict["Name"] = name
    new_dict["Caller Type"] = caller_type
    return(name, caller_type, results)
```

```python
names = []
big_list = []
for row in wrong.itertuples():
    new_dict = {}
    number = row.OFFICE_TELEPHONE
    name, caller_type, results = twilio_lookup(number)
    new_dict["Name"] = name
    new_dict["Caller Type"] = caller_type
    big_list.append(results)
    names.append(new_dict)
```

```python
bloop = pd.DataFrame(names)
bloop.to_csv('wrong_twilio.csv')
```

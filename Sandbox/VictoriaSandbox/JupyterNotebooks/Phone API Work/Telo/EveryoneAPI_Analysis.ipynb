```python
import pandas as pd
```

```python
everyone_wrong = pd.read_csv("everyone_numbers_wrong.csv")
everyone_connected = pd.read_csv("everyone_numbers_connected.csv")
everyone_disconnected = pd.read_csv("everyone_numbers_disconnected.csv")
wrong = pd.read_csv("../wrong_numbers.csv")
connected = pd.read_csv("../connected.csv")
disconnected = pd.read_csv("../disconnected.csv")
```

```python
wrong_df = pd.merge(everyone_wrong, wrong, on="OFFICE_TELEPHONE")
disconnected_df = pd.merge(everyone_disconnected, disconnected, on="OFFICE_TELEPHONE")
connected_df = pd.merge(everyone_connected, connected, on="OFFICE_TELEPHONE")
```

```python
def is_healthy(name):
    med_words = ['cancer', 'cardiology', 'neurology', 'family care', 'pulmonary', 'anesthesia', 'orthope', 'urgent care', 'allergy', 'kidney', 'surgery', 'hosp', 'mri', 'throat', 'dentist', 'med', 'clinic', 'health', 'gastroenter','anesthesiologist','patient','physician','surgeon','doctor','hospital', 'md', 'medical', 'pediatrics', 'm.d.']
    healthy = False
    for word in med_words:
        if word in str(name).lower():
            healthy = True   
    return(healthy)
```

```python
def analyze_numbers(dataframe):
    
    name_match = 0
    healthy = 0
    address_match = 0
    city_match = 0
    no_address = 0
    zipcode_match = 0
    workplace_match = 0


    for row in dataframe.itertuples():
        address_matched = False
        if str(row.PHYSICIAN_LAST_NAME).lower() in str(row.NAME).lower():
            name_match +=1
        if is_healthy(row.NAME) == True:
            healthy += 1
        if str(row.OFFICE_ADDRESS_LINE_2).lower() in str(row.ADDRESS).lower():
            address_match +=1
            address_matched = True
        if str(row.OFFICE_ADDRESS_LINE_1).lower() in str(row.NAME).lower():
            workplace_match +=1
        print(f' {row.OFFICE_ADDRESS_LINE_1} is not {row.NAME}')
        if address_matched == False:
            print (f'{row.OFFICE_ADDRESS_LINE_2} is not {row.ADDRESS}')
        if str(row.OFFICE_ADDRESS_CITY).lower() in str(row.CITY).lower():
            city_match += 1
        if row.OFFICE_ADDRESS_ZIP == row.ZIP:
            zipcode_match +=1
        if row.ADDRESS == 'nope':
            no_address += 1



    new_dict = {
        'Address Matches': address_match,
        'City Matches': city_match,
        'Zipcode Matches': zipcode_match,
        'Exact Name Matches': name_match,
        'Relevant Names': healthy,
        'No Address': no_address,
        'Workplace Name Match': workplace_match
        }
        
    return(new_dict)
```

```python
wrong_df = wrong_df.fillna('nope')
analyze_numbers(wrong_df)
```

```python
connected_df =connected_df.fillna("nope")
analyze_numbers(connected_df)
```

```python
disconnected_df = disconnected_df.fillna("nope")
analyze_numbers(disconnected_df)
```

```python
def reverse_lookup_everyone(phone_number):

    params = {'account_sid': account_sid, 'auth_token': auth_token}
    base_url = f'https://api.everyoneapi.com/v1/phone/+1{phone_number}'
    response = requests.get(base_url, params=params)
    results = response.json()
    NEW_DICT = {}
    try:
        NEW_DICT['ADDRESS'] = results['data']['address']   
    except:
        NEW_DICT['ADDRESS'] = 'N/A'
    try:
        NEW_DICT['CNAM'] = results['data']['cnam']
    except:
        NEW_DICT['CNAM'] = 'N/A'
    try:
        NEW_DICT['LINETYPE'] = results['data']['linetype']
    except:
        NEW_DICT['LINETYPE'] = 'N/A'
    try:
        NEW_DICT['CITY'] = results['data']['location']['city']
    except:
        NEW_DICT['CITY'] = 'N/A'
    try:
        NEW_DICT['STATE'] = results['data']['location']['state']
    except:
        NEW_DICT['STATE'] = 'N/A'
    try:
        NEW_DICT['ZIP'] = results['data']['location']['zip']
    except:
        NEW_DICT['ZIP'] = 'N/A'
    try:
        NEW_DICT['NAME'] = results['data']['name']
    except:
        NEW_DICT['NAME'] = 'N/A'  
    try:
        NEW_DICT['TYPE'] = results['type']
    except:
        NEW_DICT['TYPE'] = 'N/A'  
    
    return(NEW_DICT, results)
```

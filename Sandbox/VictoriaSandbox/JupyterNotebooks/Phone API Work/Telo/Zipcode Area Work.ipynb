```python
def get_zipcode_land_area(zipcode):
    url = f'https://www.zip-codes.com/zip-code/{zipcode}/zip-code-{zipcode}.asp'
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    infos = soup.find_all(class_="info")
    titles = soup.find_all('tr')
    count = 0
    for title in titles:
        if title.find('td').text =='Land Area:':
            land_area = title.find(class_ = 'info').text
            break
    return(land_area)
```

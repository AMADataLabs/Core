```python
def reverse_CNAM(number):
    params = {'account_sid': account_sid, 'auth_token': auth_token, 'format': 'json'}
    base_url = f'https://api.opencnam.com/v3/phone/+1{number}?'

    response = requests.get(base_url, params=params)
    results = response.json()
    try:
        name = results['name']
    except:
        name = 'N/A'
    
    return(name)
```

```python
def get_CNAM(df):
    dict_list = []
    for row in df.itertuples():
        new_dict = {}
        try:
            name = reverse_CNAM(row.OFFICE_TELEPHONE)
        except:
            name = "N/A"
        new_dict['Name_CNAM'] = name
        new_dict['OFFICE_TELEPHONE'] = row.OFFICE_TELEPHONE
        dict_list.append(new_dict)
    return(dict_list)
```

```python
connected_CNAM = get_CNAM(connected_df)
```

```python
connected_CNAM_df = pd.DataFrame(connected_CNAM)
connected_OpenCNAM_df = pd.merge(connected_CNAM_df, connected_df, on='OFFICE_TELEPHONE')
connected_OpenCNAM_df.head()
```

```python
disconnected_CNAM = get_CNAM(disconnected_df)
```

```python
def middle_match(name_1,name_2):
    if name_1!='None' and name_2!='None':
        if name_1[0] == name_2[0]:
            return True
        else:
            return False
    elif name_1=='None' and name_2=='None':
        return True
    else:
        return False
```

```python
def spec_match(spec_1,spec_2):
    match = False
    if spec_1 == spec_2:
        match = True
    elif spec_1 in spec_2 or spec_2 in spec_1:
        match = True
    elif fuzz.ratio(spec_1, spec_2) > 90:
        match = True
    else:
        word_count = 0
        spec_split = spec_1.split(' ')
        for word in spec_split:
            if word == 'ORTHOPEDICS':
                if 'ORTHOPAEDIC' in spec_2:
                    match = True
            if word in spec_2:
                word_count += 1
                if len(word)>8:
                    match = True
        if word_count/len(spec_split) > 0.5:
            match = True
    return match
```

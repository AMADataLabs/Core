```python
import pandas as pd
```

```python
data = pd.read_csv('../../Data/Martketing/cpt_quarterly.csv')
```

```python
data.drop_duplicates('Email')
```

```python
data.columns
```

```python
data = data.fillna('None')
```

```python
data['EMPPID'] = [int(x) if x != 'None' else 'None' for x in data.EMPPID]
data['BUSPID'] = [int(x) if x != 'None' else 'None' for x in data.BUSPID]
data['day_phone'] = [int(x) if x != 'None' else 'None' for x in data.day_phone]
data['evening_phone'] = [int(x) if x != 'None' else 'None' for x in data.evening_phone]
```

```python
emails = []
dicto = {}
```

```python
data_dict = data.fillna('None').to_dict('list')
```

```python
index = 0
for row in data.itertuples():
    print(row.Email)
    if row.Email in emails:
        for col in list(data.columns):
            if dicto[row.Email][col] == ['None']:
                dicto[row.Email][col] = [data_dict[col][index]]
            elif data_dict[col][index] not in dicto[row.Email][col] and data_dict[col][index] != 'None':
                dicto[row.Email][col].append(data_dict[col][index])
    else:
        emails.append(row.Email)
        dicto[row.Email] = {}
        for col in list(data.columns):
            dicto[row.Email][col] = [data_dict[col][index]]
    index += 1
```

```python
pd.DataFrame(dicto).transpose().to_csv('../../Data/Martketing/CPT_quarterly_squashed_2.csv', index=False)
```

```python
for email in emails:
    for thing in dicto[email]:
        if len(dicto[email][thing]) == 1:
            dicto[email][thing] = dicto[email][thing][0]
```

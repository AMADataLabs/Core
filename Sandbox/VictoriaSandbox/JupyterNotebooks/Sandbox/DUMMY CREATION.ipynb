```python
import pandas as pd
import numpy as np
```

```python
practice = pd.read_csv('../../Data/Measurement/CREDENTIALS_DUMMY.csv')
practice = practice[['Universe', 'Data Element', 'Numerator', 'Universe Total', 'Date',
       'Measure', 'Credential']]
practice_dict = practice.to_dict('records')
values = practice.Numerator/practice['Universe Total']
dates = ['5/6/2022','5/27/2022','6/10/2022','7/1/2022','7/22/2022']
```

```python
practice_list = []
for date in dates:
    new_percent = np.random.uniform(0.9, 1.1, 1)[0]
    for row in practice_dict:
        new_total = round(row['Universe Total']*new_percent)
        if row['Universe Total']!=0:
            new_value = (row['Numerator']/row['Universe Total'])-np.random.uniform(-.1, .1, 1)[0]
        else:
            new_value = 0
        new_numerator = round(new_total*new_value)
        if row['Numerator'] == 0:
            new_numerator = 0
        if new_numerator < 0:
            chunk = ((row['Numerator']/row['Universe Total'])*.1)+1
            new_numerator = round(row['Numerator']*chunk)
        if new_numerator > new_total:
            new_numerator = new_total
        new_dict = {'Universe': row['Universe'],
                   'Data Element': row['Data Element'],
                    'Numerator': new_numerator,
                    'Universe Total': new_total,
                    'Date': date,
                    'Measure': row['Measure'],
                   'Credential':row['Credential']}
        practice_list.append(new_dict)
```

```python
pd.DataFrame(practice_list+practice_dict)
```

```python
pd.DataFrame(practice_list+practice_dict).to_csv('../../Data/Measurement/CREDENTIAL.csv', index=False)
```

### Read and aggregate result data

```python
result_dataframes = []
for file in result_files:
    data = pd.read_excel(file, dtype=str)
    result_dataframes.append(data.drop_duplicates())
```

```python
result_data = pd.concat(result_dataframes, ignore_index=True)

result_data.head()
```

```python
sample_dataframes = []
for file in sample_files:
    data = pd.read_excel(file, dtype=str)
    sample_dataframes.append(data.drop_duplicates())
```

```python
sample_data = pd.concat(sample_dataframes, ignore_index=True)
sample_data.head()
```

```python
sample_data['SAMPLE_ID'].value_counts()  # verifying the Masterfile samples are here
```

```python
# filter to relevant sample phone data and rename phone column for clarity when comparing later
sample_df = sample_data[
    ['SAMPLE_ID', 'ROW_ID', 'ME', 'TELEPHONE_NUMBER']
].drop_duplicates().rename(
    columns={
        'TELEPHONE_NUMBER': 'SAMPLE_PHONE'
    }
)
```

```python
# same for result data
result_df = result_data[
    ['PHYSICIAN ME NUMBER', 'OFFICE TELEPHONE', 
     'OFFICE PHONE VERIFIED/UPDATED', 'COMMENTS', 
     'SOURCE', 'SAMPLE_ID', 'ROW_ID']
].drop_duplicates().rename(
    columns={
        'PHYSICIAN ME NUMBER': 'ME', 
        'OFFICE TELEPHONE': 'RESULT_PHONE', 
        'OFFICE PHONE VERIFIED/UPDATED': 'UPDATED'
    }

)
```

```python
sample_df.head()
```

```python
result_df.head()
```

```python
# merge
```

```python
df = sample_df.merge(result_df)  # auto-merges on shared columns--ME, SAMPLE_ID, ROW_ID
```

```python
df
```

```python
result_df.shape  # only missing a bit of result data from IQVIA, Symphony, NT samples
```

```python
# check for updates of phone
```

```python
df['UPDATED_FROM_ORIGINAL'] = df['SAMPLE_PHONE'].apply(str.strip) != df['RESULT_PHONE'].apply(str.strip)
```

### isolated data for which we need to end-date phone:
1. original phone updated

2. comment falls under 'to_end' category

```python
df['COMMENTS'].value_counts()
```

```python
to_end = [
    'MOVED, NO FORWARDING INFO',
    'NOT IN SERVICE',
    'RETIRED',
    'WRONG NUMBER',
    'FAX MODEM',
    'DECEASED',
    'DO NOT CALL',
    'REFUSAL'
]
```

```python
# updated data
to_end_updated = df[df['UPDATED_FROM_ORIGINAL'] == True][['ME', 'SAMPLE_PHONE']]

# confirmed bad results
to_end_incorrect = df[df['COMMENTS'].isin(to_end)][['ME', 'SAMPLE_PHONE']]
```

```python
to_end_updated
```

```python
to_end_incorrect
```

```python
data_to_end = pd.concat([to_end_updated, to_end_incorrect], ignore_index=True).drop_duplicates()

data_to_end
```

### with data to end-date isolated and aggregated, we can make batch file

```python
batch_file_df = pd.DataFrame()
batch_file_df['me'] = data_to_end['ME']
batch_file_df['phone'] = data_to_end['SAMPLE_PHONE']
batch_file_df['fax'] = ''
batch_file_df['source'] = ''
batch_file_df['verified_date'] = ''
batch_file_df['load_type'] = 'D'
```

```python
batch_file_df
```

```python
from datetime import datetime

today = datetime.now().date()

save_filename = f"HSG_PHYS_DELETES_{today}_StdKnownBad.csv"
save_filename
```

```python
batch_file_df.to_csv(save_filename, index=False)
```

# Validation

The validation survey processing outputs TWO batch load files:
1. batch load end-date data
2. batch load update data 

For the end-date data, basically use the same strategy as for standard survey but with a different raw result file format.

Raw files are uploaded here:

**"U:/Source Files/Data Release/Sandbox/WSLive Validation Returns/"**

As we can use a different pattern to identify the validation result file, simply add the result file we want to process to the same "result_files" folder for this project.

```python
validation_file = glob('result_files/VALIDATIONS*.xlsx')[0]  # should only be 1, ensure it's the right one 

validation_file
```

```python
sheet_names = ['ValidatedCorrect', 'ValidatedIncorrect', 'NotValidated', 'Unfinalized']

val_result_dataframes = []

for sheet in sheet_names:
    data = pd.read_excel(validation_file, sheet_name=sheet, dtype=str)
    val_result_dataframes.append(data)
    
val_result_data = pd.concat(val_result_dataframes, ignore_index=True)

val_result_data
```

```python
val_result_df = val_result_data[
    ['SAMPLE_ID', 'ROW_ID', 'ME_NO', 'OFFICE_PHONE', 'ORIGINAL_PHONE', 'CORRECT_PHONE', 'CAPTURED_NUMBER']
].drop_duplicates()
```

```python
val_result_df
```

```python
# replacement file
val_update_df = pd.DataFrame()

val_update_df['me'] = val_result_df['ME_NO']
val_update_df['phone'] = val_result_df['CAPTURED_NUMBER']
val_update_df['fax'] = ''
val_update_df['source'] = ''
val_update_df['verified_date'] = ''
val_update_df['load_type'] = 'R'  # 'replace'

val_update_df = val_update_df[~val_update_df['phone'].isna()]

val_update_df
```

```python
# each row in the replacement file corresponds to an ORIGINAL phone that should be end-dated
val_update_end_df = pd.DataFrame()

val_update_end_df['me'] = val_result_df.loc[~val_result_df['CAPTURED_NUMBER'].isna(), 'ME_NO']
val_update_end_df['phone'] = val_result_df.loc[~val_result_df['CAPTURED_NUMBER'].isna(), 'ORIGINAL_PHONE']
val_update_end_df['fax'] = ''
val_update_end_df['source'] = ''
val_update_end_df['verified_date'] = ''
val_update_end_df['load_type'] = 'D'  # 'delete'

val_update_end_df = val_update_end_df[~val_update_end_df['phone'].isna()]

val_update_end_df
```

```python
# now for the data to end-date

val_end_df = val_result_df[val_result_df['CORRECT_PHONE'].fillna('').apply(str.strip) == 'N']
val_end_df
```

```python
val_end_incorrect_df = pd.DataFrame()
val_end_incorrect_df['me'] = val_result_df.loc[val_result_df['CORRECT_PHONE'].fillna('').apply(str.strip) == 'N', 'ME_NO']
val_end_incorrect_df['phone'] = val_result_df.loc[val_result_df['CORRECT_PHONE'].fillna('').apply(str.strip) == 'N', 'ORIGINAL_PHONE']
val_end_incorrect_df['fax'] = ''
val_end_incorrect_df['source'] = ''
val_end_incorrect_df['verified_date'] = ''
val_end_incorrect_df['load_type'] = 'D'  # 'delete'

val_end_incorrect_df
```

```python
# combine the validation data to end-date (original phones which were updated and the incorrect ones)

val_end_df = pd.concat([val_update_end_df, val_end_incorrect_df], ignore_index=True)

val_end_df
```

```python
val_end_date_save_filename = f"HSG_PHYS_PHNSURV_PHONE_{today}_ValKnownBad.csv"

val_end_date_save_filename
```

```python
val_end_df.to_csv(val_end_date_save_filename, index=False)
```

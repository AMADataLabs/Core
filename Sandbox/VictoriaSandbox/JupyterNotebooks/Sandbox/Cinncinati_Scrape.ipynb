```python
cinc = pd.read_csv('../../Data/Rosters/cincinnati_emails.csv')
```

```python
link_list = list(cinc.Link)
```

```python
link_list
```

```python
for link in link_list[1422:]:
    if 'None' in link:
        continue
    response = requests.get(link)
    soup = BeautifulSoup(response.content, 'html.parser')
    title_split = soup.title.text.split(', ')
    if len(title_split)>1:
        degree = title_split[1]
    else:
        degree = 'None'
    titles = soup.find_all(class_='title')
    institutions = soup.find_all(class_='institution')
    departments = ''
    roles = ''
    depts = ''
    for thing in titles:
        roles += f'{thing.text}/'
    for dept in institutions:
        depts += f'{dept.text}/'
    new_dict ={
        'DEGREE':degree,
        'TITLE': roles[:-1],
        'DEPARTMENT': depts[:-1],
        'LINK':link
    }
    dict_list.append(new_dict) 
    print(new_dict)
```

```python
df_analysis = df[~df['COMMENTS'].isin(['DUPLICATE'])]
```

```python
inconclusive_comments = [
    'REFUSAL',
    '2ND ATTEMPT',
    'ANSWERING SERVICE',
    'LANGUAGE/HEARING',
    'DO NOT CALL'
]
```

```python
known_bad_comments
```

```python
def get_phone_status(comment, orig_phone_updated):
    #comment = row['COMMENTS']
    #updated = row['orig_phone_updated']
    
    if orig_phone_updated:
        return 'UPDATED'
    
    if comment in known_bad_comments:
        return 'KNOWN BAD'
    
    if comment in inconclusive_comments:
        return 'INCONCLUSIVE'
    
    if comment in ['COMPLETE', 'RESPONDED TO SURVEY - AMA'] and not orig_phone_updated:
        return 'CORRECT'
    
    return 'NOT-MAPPED'
```

```python
df_analysis['PHONE_STATUS'] = [get_phone_status(c, u) for c, u in zip(df_analysis['COMMENTS'].values, 
                                                                      df_analysis['orig_phone_updated'].values)
                              ]
```

```python
df_analysis['PHONE_STATUS'].value_counts() / len(df_analysis)
```

```python
df_analysis_exc_unk = df_analysis[df_analysis['PHONE_STATUS'] != 'INCONCLUSIVE']

df_analysis_exc_unk['PHONE_STATUS'].value_counts() / len(df_analysis_exc_unk)
```

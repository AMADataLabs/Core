```python
countries = pd.read_csv('../../Data/Measurement/these_ones.csv')
```

```python
import numpy as np
  
# function to get unique values
def unique(list1):
    x = np.array(list1)
    print(np.unique(x))
```

```python
dict_list = []
bad_countries = []
for row in countries.itertuples():
    print('')
    print(f'{row.COUNTRY_CD} : {row.DESC}')
    try:
        numeric = pycountry.countries.lookup(row.COUNTRY_CD).alpha_3
        new_dict = {
            'alpha_3':numeric,
            'COUNTRY_ID':row.COUNTRY_ID
        }
        dict_list.append(new_dict)
        continue
    except LookupError:
        pass
    try:
        numeric = pycountry.countries.lookup(row.DESC).alpha_3
        new_dict = {
            'alpha_3':numeric,
            'COUNTRY_ID':row.COUNTRY_ID
        }
        dict_list.append(new_dict)
        continue
    except LookupError:
        pass
    try:
        numeric = pycountry.historic_countries.lookup(row.COUNTRY_CD).alpha_3
        new_dict = {
            'alpha_3':numeric,
            'COUNTRY_ID':row.COUNTRY_ID
        }
        dict_list.append(new_dict)
        continue
    except LookupError:
        pass
    try:
        numeric = pycountry.historic_countries.lookup(row.DESC).alpha_3
        new_dict = {
            'alpha_3':numeric,
            'COUNTRY_ID':row.COUNTRY_ID
        }
        dict_list.append(new_dict)
        continue
    except LookupError:
        bad_countries.append(row.COUNTRY_ID)
        print('Country not found')
```

```python
pd.DataFrame(dict_list)
```

```python
countries[countries.COUNTRY_ID.isin(bad_countries)].to_csv('../../Data/Measurement/bad_countries.csv', index=False)
```

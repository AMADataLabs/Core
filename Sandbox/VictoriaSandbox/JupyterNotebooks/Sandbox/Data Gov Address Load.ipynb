```python
cols = ['entity_id',
                'me#',
                'comm_id',
                'usage',
                'load_type',
                'addr_type',
                'addr_line_1',
                'addr_line_2',
                'addr_line_3',
                'addr_city',
                'addr_state',
                'addr_zip',
                'addr_plus4',
                'addr_country',
                'source',
                'source_dtm'
                ] 
def create(gov):
    today = str(date.today())
    col_rename = {
            'adr_ln_1':'addr_line_1',
            'cty':'addr_city',
            'st':'addr_state',
            'adr_ln_2':'addr_line_2',
            'ME':'me#'
        }
    gov = gov.rename(columns=col_rename)
    cols = ['entity_id',
                'me#',
                'comm_id',
                'usage',
                'load_type',
                'addr_type',
                'addr_line_1',
                'addr_line_2',
                'addr_line_3',
                'addr_city',
                'addr_state',
                'addr_zip',
                'addr_plus4',
                'addr_country',
                'source',
                'source_dtm'
                ]
    gov['addr_zip'] = [use.fix_zipcode(x) for x in gov.zip]
    gov['addr_plus4'] = [x if len(str(x))==9 else '' for x in gov.zip]
    gov['load_type'] = 'A'
    gov['addr_type'] = 'OF'
    gov['source'] = 'NAT-DOWN'
    gov['source_dtm'] = today
    gov['entity_id'] = ''
    gov['comm_id'] = ''
    gov['usage'] = ''
    gov['addr_country'] = ''
    gov['addr_line_3'] = ''
    gov = gov.drop_duplicates(['me#','adrs_id'])[cols]
    return (gov)
```

```python
gov.drop_duplicates(['me#','adrs_id'])[cols].to_csv('../../Data/DataGov/address_load_national_downloadable_2021-11-15.csv',index=False)
```

```python
home_words = [' APT','APARTMENT']
office_words = ['SUITE','STE']
ppd = ppd.fillna('None')
should_be_home = []
should_be_office = []
for row in ppd.itertuples():
    if row.ADDRESS_TYPE == 1.0:
        for word in home_words:
            if word in row.MAILING_LINE_2 or word in row.MAILING_LINE_1:
                mes.append(row.ME)
                break
    if row.ADDRESS_TYPE == 2.0:
        for word in office_words:
            line_words = row.MAILING_LINE_2.split(' ')+(row.MAILING_LINE_1.split(' '))
            for other_word in line_words:
                if word == other_word:
                    mes.append(row.ME)
                    break
```

```python
import pandas as pd
```

```python
ppd = pd.read_csv('../../Data/Scorecard/ppd_data_20211106.csv')
```

```python
dpc = ppd[ppd.TOP_CD==20]
```

```python
polo_columns = ['POLO_MAILING_LINE_1', 'POLO_MAILING_LINE_2','POLO_CITY', 'POLO_STATE', 'POLO_ZIP']
```

```python
dpc = dpc.fillna('None')
```

```python
this_dict_list = []
for column in polo_columns:
    count = len(dpc[dpc[column]!='None'])
    this_dict = {
        'COLUMN':column,
        'COUNT':count
    }
    this_dict_list.append(this_dict)
```

```python
pd.DataFrame(this_dict_list).to_csv('../../Data/Scorecard/POLO_Count_Options.csv', index=False)
```

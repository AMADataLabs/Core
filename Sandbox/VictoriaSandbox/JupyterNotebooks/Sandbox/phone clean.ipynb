```python
dict_list = []
for row in cities.itertuples():
    spliz = row.Code.split('-')
    area_code = spliz[0].strip()
    location = spliz[1].split(',')
    if len(location)==2:
        city = location[0].strip()
        state = state_codes[location[1].strip()]
    else:
        state = spliz[1].strip()
        city = None
    new_dict = {
        'area_code':area_code,
        'state':state,
        'city':city
    }
    dict_list.append(new_dict)
```

```python
dict_list = []
for row in states.itertuples():
    spliz = row.Code.split(',')
    area_code = spliz[0].strip()
    location = spliz[1].split(',')
    if len(location)==2:
        city = location[0].strip()
        state = state_codes[location[1].strip()]
    else:
        state = spliz[1].strip()
        city = None
    new_dict = {
        'area_code':area_code,
        'state':state,
        'city':city
    }
    dict_list.append(new_dict)
```

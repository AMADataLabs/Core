```python
import pandas as pd
```

```python
results_1 = pd.read_csv('../../Data/PhoneAppend/Bing_Results_1.csv')
results_2 = pd.read_csv('../../Data/PhoneAppend/Bing_Search_Results.csv')
og_sample = pd.read_csv('../../Data/PhoneAppend/BingSample.csv')
state_area = pd.read_csv('../../Data/PhoneAppend/state_area_codes.csv')
```

```python
results_3 = pd.read_csv('../../Data/PhoneAppend/Bing_Results_final.csv')
```

```python
def only_numerics(seq):
    seq_type= type(seq)
    return seq_type().join(filter(seq_type.isdigit, seq))

def is_repeating(number):
    previous_digit = str(number)[0]
    for digit in str(number):
        if digit != previous_digit:
            return(False)
        else:
            previous_digit = digit
    return(True)

def clean_phones(unique_phones):
    num_list =[]
    for number in unique_phones.NUMBERS:
        new_number = number
        if len(number)<10:
            new_number = 'None'
        elif len(number)>10:
            if number[0]=='1':
                new_number = number[1:]
            new_number = only_numerics(new_number)
        if len(new_number)!=10 or is_repeating(new_number):
            new_number = 'None'
        num_list.append(new_number)
    unique_phones['OFFICE_PHONE'] = num_list
    return(unique_phones)

def check_area_codes(sample, area_codes):
    sample_area = pd.merge(sample, area_codes, on='STATE')
    matches = []
    for row in sample_area.itertuples():
        match = False
        codes = str(row.Areacodes).split(',\xa0')
        toll_free_codes = ['800', '888', '877', '866', '855', '844', '833']
        if str(row.OFFICE_PHONE)[0:3] in codes:
            match = True
        elif str(row.OFFICE_PHONE)[0:3] in toll_free_codes:
            match = True
        matches.append(match)
    sample_area['AREA_CODE_MATCH']=matches
    return(sample_area)
```

```python
all_results = pd.concat([results_1, results_2, results_3])
results = pd.merge(all_results, og_sample, left_on='NAME', right_on='MAILING_NAME')
```

```python
all_results = all_results[['NAME','NUMBERS', 'URL']]
```

```python
all_results = clean_phones(all_results)
```

```python
phone_results = all_results[all_results.OFFICE_PHONE!='None']
```

```python
phone_results = pd.merge(og_sample[['MAILING_NAME','STATE']], phone_results, left_on='MAILING_NAME', right_on='NAME')
```

```python
phone_results = check_area_codes(phone_results, state_area)
```

```python
phone_results = phone_results[phone_results.AREA_CODE_MATCH==True]
```

```python
phone_results.groupby(['NAME','OFFICE_PHONE']).count().sort_values('STATE', ascending=False)[['MAILING_NAME']].to_csv('../../Data/PhoneAppend/Best_Phones_May.csv')
```

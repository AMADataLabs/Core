### Individual has no phone, but does have a POLO
#### For the scenario where the individual has a POLO, we could utilize that address as one of the data points in acquiring a phone that weâ€™re comfortable saying is appropriate. This might be best served by the Bing search methodology.

```python
def is_a_match(thing_1, thing_2):
    if thing_1 == thing_2:
        match = True
    elif thing_1 in thing_2:
        match = True
    elif thing_2 in thing_1:
        match = True
    elif fuzz.ratio(thing_1, thing_2)>80:
        match = True
    else:
        match = False
    return(match)
```

```python
just_a_test = pd.merge(yes_polo, gov, on='ME')
```

```python
dicts =[]
for row in just_a_test.itertuples():
    if row.POLO_STATE == row.st:
        if is_a_match(row.adr_ln_1, row.POLO_MAILING_LINE_2):
            print(f'{row.zip} is {row.POLO_ZIP}')
            new_dict = {
                'ME': row.ME,
                'adrs_id': row.adrs_id,
                'STATE': row.STATE,
                'GOV_ADDR': row.adr_ln_1,
                'POLO': row.POLO_MAILING_LINE_2
                }
            dicts.append(new_dict)
```

```python
pd.DataFrame(dicts).drop_duplicates(['ME'])
```

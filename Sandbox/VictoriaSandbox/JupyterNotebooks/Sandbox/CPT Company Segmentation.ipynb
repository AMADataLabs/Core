```python
#import library
import pandas as pd
#reads page one of the excel doc you sent me into a pandas dataframe
data_lic = pd.read_excel('../../Data/CPT_Company_Segmentation/database_lic_score.xlsx')
#creates a list of the relevant columns
key_list = ['billing', 'automation', 'ai', 'cycle', 'health',
       'healthcare', 'predictive', 'data', 'drug', 'diagnostic',
       'optimization', 'claims', 'therapeutics', 'analytics', 'artificial',
       'cancer', 'biopharma', 'diabetes', 'dental', 'reimbursement',
       'machine']
#creates a smaller dataframe with only these columns
keywords = data_lic[key_list]
#converts the dataframe to a list of python dictionaries
key_dict = keywords.to_dict('records')
#creates an empty list
dict_list = []
#iterates through each "row" in the dictionary list and creates a dictionary for every company-keyword pair if the column is true for that keyword in that row
for row in key_dict:
    for keyword in key_list:
        if row[keyword]:
            new_dict = {'Companies': row['Companies'],
                       'Keyword': keyword}
            dict_list.append(new_dict)
#creates a csv with the new data
pd.DataFrame(dict_list).to_csv('../../Data/CPT_Company_Segmentation/test.csv', index=False)
```

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
def scrape_countries(df):
    base_url = 'https://en.wikipedia.org/wiki/'
    countries=[]
    bad_cities =[]
    for row in df.itertuples():
        print('--------------------')
        city = row.BIRTH_CITY.title()
        citi = city.replace(" ","_")
        url = base_url + citi
        response = requests.get(url)
        if response.status_code == 200:
            soup = BeautifulSoup(response.content, 'html.parser')
            test = soup.find_all(class_="mergedtoprow")
            for thing in test:
                x = thing.find_all('td')
                y = thing.find_all('th')
                for this in y:
                    if this.text == 'Country':
                        country = x[0].text
                        country=country.replace('\xa0','')
                        print(f'City: {city}\nCountry: {country}')
        else:
            bad_cities.append(city)
            country='None'
            print(f'{city} needs human intervention')
        countries.append(country)
        print("")
    return(countries, bad_cities)
```

```python
country_work = pd.read_csv("AMAIA/Country_Work.csv")
```

```python
COUNTRY = country_work[(country_work['countrycode']=='None') & (country_work['BIRTH_CITY']!='UNKNOWN')]
COUNTRU = COUNTRY.drop_duplicates(subset='BIRTH_CITY')
```

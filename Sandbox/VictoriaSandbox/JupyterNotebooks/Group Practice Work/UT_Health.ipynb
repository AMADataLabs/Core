```python
dict_list =[]
for provider in providers:
    name = provider.find(class_='provider-title').text.replace('\n','')
    link = provider.find('a')['href']
    specialty = str(provider.find('p')).replace('<p>','').split('<br/>')[0]
    specialty_2 = str(provider.find('p')).replace('<p>','').split('<br/>')[1]
    if specialty_2 == '</p>':
        specialty_2 = 'None'
    phones =[]
    locations =[]
    for thing in provider.find_all('strong'):
        if thing.text[0]=='(':
            phones.append(thing.text)
        else:
            locations.append(thing.text)
    new_dict = {
        'Link': link,
        'Name': name,
        'Specialty': specialty,
        'Secondary_Specialty': specialty_2,
        'Phones': phones,
        'Location_Names':locations,
        'From_DT': dt.datetime.now()
    }
    dict_list.append(new_dict)
```

```python
nums = list(range(3,27))
nums
```

```python
for num in nums:
    url = f'https://www.utphysicians.com/providers/page/{num}/'
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    providers= soup.find_all(class_ = 'provider-listing')
    for provider in providers:
        name = provider.find(class_='provider-title').text.replace('\n','')
        link = provider.find('a')['href']
        specialty = str(provider.find('p')).replace('<p>','').split('<br/>')[0]
        if specialty == '</p>':
            specialty = 'None'
        try:
            specialty_2 = str(provider.find('p')).replace('<p>','').split('<br/>')[1]
        except:
            specialty_2 = 'None'
        if specialty_2 == '</p>':
            specialty_2 = 'None'
        phones =[]
        locations =[]
        for thing in provider.find_all('strong'):
            if len(thing.text) > 0:
                if thing.text[0]=='(':
                    phones.append(thing.text)
                else:
                    locations.append(thing.text)
        new_dict = {
            'Link': link,
            'Name': name,
            'Specialty': specialty,
            'Secondary_Specialty': specialty_2,
            'Phones': phones,
            'Location_Names':locations,
            'From_DT': dt.datetime.now()
        }
        dict_list.append(new_dict)
    print(f'Page {num} done!')
```

```python
df = pd.DataFrame(dict_list)
```

```python
df.drop_duplicates(subset = 'Name').to_csv('Ut_Health_Roster.csv')
```

```python
mayo = pd.read_csv('Mayo_Roster.csv')
```

```python
rochester = mayo[mayo['Location']=='Rochester, MN']
```

```python
dict_list =[]
for row in mayo.itertuples():
    new_dict ={}
#     name=row.Name.replace(',','').replace('.','').replace('MD','').replace('MB','').replace('CM','').replace('MS','').replace('MPH','').replace('PhD','')
    name_split = row.Name.split(",")
    name_list=name_split[0].split(' ')
    FIRST_NAME = name_list[0].upper()
    MIDDLE_NAME='None'
    ALTERNATE_NAME = 'None'
    if len(name_list[1])==2 and name_list[1][1]=='.':
        MIDDLE_NAME=name_list[1].replace('.','')
    if name_list[1][0]=='(':
        ALTERNATE_NAME = name_list[1].replace(')', '').replace('(','')
    LAST_NAME = name_list[-1].upper()
    new_dict['Name_List']=name_list
    new_dict['Name']=row.Name
    new_dict['FIRST_NAME']=FIRST_NAME
    new_dict['LAST_NAME']=LAST_NAME
    new_dict['MIDDLE_NAME']=MIDDLE_NAME
    new_dict['ALTERNATE_NAME']=ALTERNATE_NAME.upper()
    dict_list.append(new_dict)
```

```python
namess = pd.DataFrame(dict_list)
```

```python
with_names = pd.merge(rochester, namess, on='Name')
```

```python
pd.merge(rochester_address, with_names, on = ['FIRST_NAME', 'LAST_NAME'])
```

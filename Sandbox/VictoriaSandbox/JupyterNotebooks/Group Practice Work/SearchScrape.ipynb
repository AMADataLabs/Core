```python
index = list(range(332,449))
```

```python
for number in index:
    url = 'https://www.henryford.com/physician-directory/search-results?page='+str(number)
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    listings = soup.find_all(class_='listing')
    for listing in listings:
        new_dict = {}
        try:
            new_dict['PHONE'] = listing.find(class_='phones').text
        except:
            new_dict['PHONE'] = 'None'
        try:
            new_dict['SPECIALITIES'] = listing.find(class_="module-pd-specialty-list").text.replace('\nSpecialties: ','').replace('\n','')
        except:
            new_dict['SPECIALTIES'] = 'None'
        addresses = listing.find_all(class_ = 'module-pd-office-item')
        address_list =[]
        address_name_list = []
        for address in addresses:
            try:
                address_name_list.append(address.find('p').text.replace('\n',''))
            except:
                address_name_list=[]
            try:
                address_list.append(address.find('a')['href'].replace('http://maps.google.com?q=',''))
            except:
                address_list=[]
        new_dict['ADDRESSES']=address_list
        new_dict['PRACTICE_NAMES']=address_name_list
        try:
            new_dict['NAME'] = listing.find('img')['alt']
        except:
            new_dict['NAME']='None'
        dict_list.append(new_dict)
    print(f'Page {number} done!')
```

```python
url = 'https://www.henryford.com/physician-directory/search-results?page=449'
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')
listings = soup.find_all(class_='listing')
for listing in listings:
    new_dict = {}
    new_dict['PHONE'] = listing.find(class_='phones').text
    new_dict['SPECIALITIES'] = listing.find(class_="module-pd-specialty-list").text.replace('\nSpecialties: ','').replace('\n','')
    addresses = listing.find_all(class_ = 'module-pd-office-item')
    address_list =[]
    address_name_list = []
    for address in addresses:
        try:
            address_name_list.append(address.find('p').text.replace('\n',''))
        except:
            address_list=[]
        address_list.append(address.find('a')['href'].replace('http://maps.google.com?q=',''))
    new_dict['ADDRESSES']=address_list
    new_dict['PRACTICE_NAMES']=address_name_list
    new_dict['NAME'] = listing.find('img')['alt']
    dict_list.append(new_dict)
```

```python
henry_ford = pd.DataFrame(dict_list)
```

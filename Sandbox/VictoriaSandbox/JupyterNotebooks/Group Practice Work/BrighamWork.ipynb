```python
def scrape_brigham():
    dict_list = []
    index = list(range(1,15))
    for number in index:
        base_url = 'https://physiciandirectory.brighamandwomens.org/?SetSize=48&Index='+str(number)
        response = requests.get(base_url)
        soup = BeautifulSoup(response.content, 'html.parser')
        providers = soup.find_all(class_ = 'provider-holder')
        for provider in providers:
            provider_dict = {}
            provider_dict['link'] = provider.find('a')['href']
            provider_dict['index'] = provider_dict['link'].split('/')[4]
            provider_dict['name'] = provider.find(class_='nameHead').text.replace('\n','')
            try:
                provider_dict['speciality'] = provider.find(class_='listSpecialties').text.replace('\t','').replace('\r','').replace('\n','')
            except:
                provider_dict['speciality']='None'
            dict_list.append(provider_dict)
            print('---------')
            print(f'{provider_dict["name"]}: {provider_dict["speciality"]}')
        print(f'Page {number} done!')
        print('...........')
    return(dict_list)
```

```python
brigham_list = scrape_brigham()
brigham_df = pd.DataFrame(brigham_list)
brigham_df
```

```python
brigham_df.to_csv('brigham_scrape_2.csv')
```

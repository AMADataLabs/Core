```python
def scrape_massachusetts():
    potential_list = list(range(16000,21000))
    list_of_dicts = []
    for number in potential_list:
        url = f'https://www.massgeneral.org/doctors/doctor.aspx?id={number}#'
        response = requests.get(url)
        if response.status_code == 200:
            print('------------------')
            new_dict ={}
            soup = BeautifulSoup(response.content, 'html.parser')
            title_ = soup.title.text.split(' -')
            name_ = title_[0]
            try:
                department_ = title_[1].replace(' ','',1)
            except:
                department_ = 'None'
            try:
                practice_ = title_[2].replace(' ','',1)
            except:
                practice_ = 'None'
            phone_ = soup.find(class_ = 'cta__phone__number').text.replace(' ','').replace('-','')
            new_dict['Name'] = name_
            new_dict['Department'] = department_
            new_dict['Practice'] = practice_
            new_dict['Telephone'] = phone_
            list_of_dicts.append(new_dict)
            print(f'Name: {name_}')
            print(f'Department: {department_}')
            print(f'Practice: {practice_}')
            print(f'Phone: {phone_}')
            print('------------------')
            print('')
    df_of_docs = pd.DataFrame(list_of_dicts)
    return(df_of_docs)
```

```python
scrape_massachusetts()
```

```python
list_of_dict_2 = []
for row in mass_hospital_roster.itertuples():
    new_dict= {}
    new_dict['Name']=row.Name
    name_list = row.Name.replace(',','').replace('.','').upper().split(' ')
    if 'PHD' in name_list:
        name_list.remove('PHD')
    if 'MD' in name_list:
        name_list.remove('MD')
    new_dict['FIRST_NAME'] = name_list[0]
    if len(name_list)>2:
        new_dict['MIDDLE_NAME'] = name_list[1]
        new_dict['LAST_NAME'] = name_list[2]
    else:
        new_dict['MIDDLE_NAME'] = 'None'
        new_dict['LAST_NAME'] = name_list[1]
    list_of_dict_2.append(new_dict)
```

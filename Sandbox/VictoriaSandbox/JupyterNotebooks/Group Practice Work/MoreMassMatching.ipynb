```python
import pandas as pd
```

```python
addresses_df = pd.read_csv('Addresses_Potential_Names.csv')
```

```python
all_addresses = []
group_addresses = []
associated_addresses = {}
group_add_df = []
for row in addresses_df.itertuples():
    if row.POLO_MAILING_LINE_1 in group_addresses:
        associated_addresses[str(row.POLO_MAILING_LINE_1)].append(row.POLO_MAILING_LINE_2)
    if row.POLO_MAILING_LINE_1 in all_addresses and row.POLO_MAILING_LINE_1 not in group_addresses:
        group_addresses.append(row.POLO_MAILING_LINE_1)
        associated_addresses[str(row.POLO_MAILING_LINE_1)] = [row.POLO_MAILING_LINE_2]
        new_dict = {}
        new_dict['Name']=row.POLO_MAILING_LINE_1
        new_dict['Address']=row.POLO_MAILING_LINE_2
        group_add_df.append(new_dict)
    elif row.ME > 1  and row.POLO_MAILING_LINE_1 not in group_addresses:
        group_addresses.append(row.POLO_MAILING_LINE_1)
        associated_addresses[str(row.POLO_MAILING_LINE_1)] = [row.POLO_MAILING_LINE_2]
        new_dict = {}
        new_dict['Name']=row.POLO_MAILING_LINE_1
        new_dict['Address']=row.POLO_MAILING_LINE_2
        group_add_df.append(new_dict)
    if row.POLO_MAILING_LINE_1 not in all_addresses:
        all_addresses.append(row.POLO_MAILING_LINE_1)
```

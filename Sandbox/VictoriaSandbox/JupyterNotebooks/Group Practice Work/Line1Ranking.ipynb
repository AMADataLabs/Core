```python
NAMES = names.dropna(subset=['POLO_MAILING_LINE_1'])
name_work = NAMES.drop_duplicates(subset=['POLO_MAILING_LINE_2','POLO_MAILING_LINE_1'])
```

```python
name_work.to_csv('name_work.csv')
```

```python
ZEE = NAMES.groupby('POLO_MAILING_LINE_1').count().sort_values('POLO_MAILING_LINE_2', ascending=False)
```

```python
ZEE.to_csv('blegh.csv')
blegh = pd.read_csv('blegh.csv')
```

```python
RANKING = blegh[['POLO_MAILING_LINE_1', 'POLO_MAILING_LINE_2']]
ranking = RANKING.rename(columns={'POLO_MAILING_LINE_2':'LINE_1_COUNT'})
```

```python
tea = pd.merge(ranking, name_work, on='POLO_MAILING_LINE_1')
```

```python
HOT_TEA = tea.drop_duplicates(subset=['POLO_MAILING_LINE_2'], keep='first')
```

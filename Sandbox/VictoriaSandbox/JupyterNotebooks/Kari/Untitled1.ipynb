```python
def clean_ent_comm_data(ent_comm_df):
    
    ent_comm_cols = ['entity_id', 'comm_type', 'begin_dt', 'comm_id', 'end_dt', 'src_cat_code']    
    ent_comm_df = ent_comm_df[ent_comm_cols]
    ent_comm_df = rename_comm_cols(ent_comm_df)
    
    return ent_comm_df
```

```python
ent_comm_df = clean_ent_comm_data(ent_comm_df)    
```

```python
def rename_comm_cols(data_df):
    col_names = data_df.columns.values
    new_col_dict = {}
    for name in col_names:
        if name.find('ent_comm') >= 0:
            new_col_dict[name] = name
        else:
            new_name = 'ent_comm_' + name
            new_col_dict[name] = new_name
            
    data_df = data_df.rename(columns = new_col_dict)
    
    return data_df
```

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://baycare.org/doctors?&page=0&pageSize=3600&returnWildcardResults=true&useDefaultLocation=false'
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
providers = soup.find_all(class_='provider')
```

```python
dict_list = []
for provider in providers:
    name = provider.find('img')['alt']
    if len(provider.find_all(class_='provider-address'))==3:
        address = provider.find_all(class_='provider-address')[0].text
        city = provider.find_all(class_='provider-address')[2].text
    else:
        print('Multiple addresses')
    for x in provider.find_all('a'):
        if 'tel' in x['href']:
            phone = x.text
    specialties = provider.find(class_='provider-specialties').text.replace('Specialties: ','')
    locations = provider.find(class_='provider-locations').text.replace('Locations in: ','')                          
    new_dict={
        'Name':name,
        'Address':address,
        'City':city,
        'Phone':phone,
        'Specialties':specialties,
        'Locations': locations
    }
                                 
    dict_list.append(new_dict)
```

```python
baycare_df = pd.DataFrame(dict_list)
```

```python
baycare_df
```

```python
new_dict_list = []
for entry in dict_list:
    name_list = clean_names(entry['Name'])
    first_name = name_list[0]
    middle_name = name_list[1]
    last_name = name_list[2]
    suffix = name_list[3]
    city_list = entry['City'].split(', ')
    city = city_list[0]
    state = city_list[1].split(' ')[0]
    zipcode = city_list[1].split(' ')[1]
    new_entry = {
        'First':first_name,
        'Last': last_name,
        'Middle': middle_name,
        'Suffix': suffix,
        'City': city,
        'State': state,
        'Zipcode': zipcode,
        'Name': entry['Name']
    }
    new_dict_list.append(new_entry)
```

```python
def clean_names(name):
    split_name = name.split(' ')
    first_name = split_name[0]
    if 'PhD' in split_name:
        split_name.remove('PhD')
    if 'MD' in split_name:
        split_name.remove('MD')
    if 'DO' in split_name:
        split_name.remove('DO')
    if len(split_name[1])==2:
        middle_name = split_name[1]
    else:
        middle_name = 'None'
    if 'Jr' in split_name:
        suffix = 'Jr'
        split_name.remove('Jr')
    else:
        suffix = 'None'
    last_name = split_name[-1]
    last_name = last_name.replace(',','')
    return([first_name, middle_name, last_name, suffix])
```

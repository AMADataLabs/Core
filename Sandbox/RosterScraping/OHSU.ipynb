```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://endeca.ohsu.edu/find-a-doctor/?term=&language=&gender=&specialty='
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
index = list(range(2,239))
```

```python
for number in index:
    url = "https://endeca.ohsu.edu/find-a-doctor/?&term=&accepting=&gender=&language=&specialty=&page=" + str(number)
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    results = soup.find_all(class_='vcard')
    for result in results:
        result_dict = {}
        result_dict['link'] = result.find('a')['href']
        result_dict['name'] = result.find('img')['alt']
        try:
            result_dict['specialities']= result.find_all('p')[1].text
        except:
            result_dict['specialities'] = 'None'
        try:
            result_dict['focus'] = result.find_all('p')[2].text.replace('Special focus in ','')
        except:
            result_dict['focus']= 'None'
        dict_list.append(result_dict)
    print(f'Page {number} done!')
    print(f'{number/239*100}% complete')
```

```python
ohsu_df = pd.DataFrame(dict_list)
```

```python
ohsu_df.to_csv('OHSU_df.csv')
```

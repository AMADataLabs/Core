```python
dict_list =[]
for result in results:
    name = result.find(class_ = 'doctor-result-title-link').text.replace('\t','').replace('\n','')
    link = result.find(class_ = 'doctor-result-title-link')['href']
    phone = result.find(class_ = 'doctor-result-phone-number').text.replace('\t','').replace('\n','')
    specialty = result.find(class_='doctor-result-specialties').text.replace('\t','').replace('\n','').replace('Specialties: ','')
    try:
        locations = result.find(class_ = 'doctor-result-locations').text.replace('\t','').replace('\n','').replace('Practices at ','')
    except:
        locations = 'None'
    new_dict = {
        'Name':name,
        'Phone':phone,
        'Specialties':specialty,
        'Locations': locations,
        'Link': link
    }
    dict_list.append(new_dict)
```

```python
page_numbers = list(range(18,53))
```

```python
for page in page_numbers:
    url = f"https://www.southcoast.org/doctors/page/{page}/?search"
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    results = soup.find_all(class_ = 'doctor-result')
    for result in results:
        name = result.find(class_ = 'doctor-result-title-link').text.replace('\t','').replace('\n','')
        link = result.find(class_ = 'doctor-result-title-link')['href']
        phone = result.find(class_ = 'doctor-result-phone-number').text.replace('\t','').replace('\n','')
        try:
            specialty = result.find(class_='doctor-result-specialties').text.replace('\t','').replace('\n','').replace('Specialties: ','')
        except:
            specialty = 'None'
        try:
            locations = result.find(class_ = 'doctor-result-locations').text.replace('\t','').replace('\n','').replace('Practices at ','')
        except:
            locations = 'None'
        new_dict = {
            'Name':name,
            'Phone':phone,
            'Specialties':specialty,
            'Locations': locations,
            'Link': link
        }
        dict_list.append(new_dict)
    print(f'Page {page} done!')
```

```python
url = f"https://www.southcoast.org/doctors/?search"
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')
results = soup.find_all(class_ = 'doctor-result')
for result in results:
    name = result.find(class_ = 'doctor-result-title-link').text.replace('\t','').replace('\n','')
    link = result.find(class_ = 'doctor-result-title-link')['href']
    phone = result.find(class_ = 'doctor-result-phone-number').text.replace('\t','').replace('\n','')
    try:
        specialty = result.find(class_='doctor-result-specialties').text.replace('\t','').replace('\n','').replace('Specialties: ','')
    except:
        specialty = 'None'
    try:
        locations = result.find(class_ = 'doctor-result-locations').text.replace('\t','').replace('\n','').replace('Practices at ','')
    except:
        locations = 'None'
    new_dict = {
        'Name':name,
        'Phone':phone,
        'Specialties':specialty,
        'Locations': locations,
        'Link': link
    }
    dict_list.append(new_dict)
```

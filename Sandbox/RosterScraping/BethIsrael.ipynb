```python
dict_list =[]
for provider in providers:
    link = provider.find('a')['href']
    name = provider.find_all('a')[1].text
    specialty = provider.find(class_='listSpecialties').text.replace('\r','').replace(', \n','').replace('  ','').replace('\n','')
    phone = provider.find_all(class_='profileData')[0]['content']
    fax = provider.find_all(class_='profileData')[1]['content']
    profile = provider.find_all(class_='profileDataNoLabel')
    location = profile[0].text
    address_2 = profile[1].text
    address = profile[2].text
    if "\n" in address:
        address = 'None'
    city = provider.find(itemprop='addressLocality').text
    zipcode = provider.find(itemprop='postalCode').text
    new_dict = {
        'Link': link,
        'Name': name,
        'Specialty': specialty,
        'Phone': phone,
        'Fax': fax,
        'Location_Name':location,
        'Address_1': address,
        'Address_2': address_2,
        'City':city,
        'Zipcode':zipcode,
        'From_DT': dt.datetime.now()
    }
    dict_list.append(new_dict)
```

```python
for num in nums:
    url = f'https://findadoc.bidmc.org/?SetSize=48&Index={num}'
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    providers = soup.find_all(class_='provider-holder')
    for provider in providers:
        link = provider.find('a')['href']
        name = provider.find_all('a')[1].text
        try:
            specialty = provider.find(class_='listSpecialties').text.replace('\r','').replace(', \n','').replace('  ','').replace('\n','')
        except:
            specialty = 'None'
        try:
            phone = provider.find_all(class_='profileData')[0]['content']
        except:
            phone = 'None'
        try:
            fax = provider.find_all(class_='profileData')[1]['content']
        except:
            fax = 'None'
        profile = provider.find_all(class_='profileDataNoLabel')
        try:
            location = profile[0].text
        except:
            location = 'None'
        try:
            address_2 = profile[1].text
        except:
            address_2 = 'None'
        try:
            address = profile[2].text
        except:
            address = 'None'
        if "\n" in address:
            address = 'None'
        try:
            city = provider.find(itemprop='addressLocality').text
        except:
            city = 'None'
        try:
            zipcode = provider.find(itemprop='postalCode').text
        except:
            zipcode = 'None'
        new_dict = {
            'Link': link,
            'Name': name,
            'Specialty': specialty,
            'Phone': phone,
            'Fax': fax,
            'Location_Name':location,
            'Address_1': address,
            'Address_2': address_2,
            'City':city,
            'Zipcode':zipcode,
            'From_DT': dt.datetime.now()
        }
        dict_list.append(new_dict)
        
    print(f'Page {num} done!')
```

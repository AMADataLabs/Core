```python
def adjust_excel_column_widths(writer: pd.ExcelWriter):
    for sheetname in writer.sheets:
        sheet = writer.sheets[sheetname]
        print('sheetname:', sheetname)
        temp_data = pd.read_excel(writer, sheet_name=sheetname)
        for idx, col in enumerate(temp_data):
            series = temp_data[col]
            max_len = max((
                series.astype(str).map(len).max(), len(str(series.name))
            ))
            sheet.set_column(idx, idx, max_len)
    writer.save()
```

```python
def get_letters_between(start, end):
    # returns letters in alphabet from <start> THROUGH <end>
    return ascii_uppercase[ascii_uppercase.index(start) : ascii_uppercase.index(end)+1]
```

```python
from string import ascii_uppercase
```

```python
sheet_column_widths = {
    'ChangeFileAudit': {
        'A:A': 31
    },
    'ReportByFieldFrom SAS': {
        'A:A': 31,
        'B:B': 8,
        'C:C': 12
    },
    'ChangeByFieldCount': {
        'A:A': 37,
        'B:B': 9,
        'C:C': 14
    },
    'RecordActionExtract': {
        'A:A': 21,
        'B:B': 12
    },
    'ChangeByRecordCount': {
        'A:A': 21,
        'C:C': 12
    },
    'PE Counts': {
        'A:A': 12,
        'B:B': 50,
        'C:C': 7
    },
    'TOP Counts': {
        'A:A': 12,
        'B:B': 29,
        'C:C': 7
    },
    'TOP by PE': {
        'A:A': 13,
        'B:B': 29,
        'U:U': 12,
    },
    'PrimSpecbyMPA': {
        'A:A': 11,
        'B:B': 80,
        'N:N': 12
    },
    'SecSpecbyMPA': {
        'A:A': 11,
        'B:B': 80,
        'N:N': 12
    }
}
for col in get_letters_between('C', 'T'):  # 'C' through 'T'
    colname = f'{col}:{col}'
    sheet_column_widths['TOP by PE'][colname] = 6
for col in get_letters_between('C', 'M'):  # 'C' through 'T'
    colname = f'{col}:{col}'
    sheet_column_widths['PrimSpecbyMPA'][colname] = 6
    sheet_column_widths['SecSpecbyMPA'][colname] = 6
```

```python
sheet_percentage_columns = {
    'ReportByFieldFrom SAS': ['C:C'],
    'ChangeByRecordCount': ['C:C']
}
```

```python
for sheetname in sheet_column_widths:
    for column in sheet_column_widths[sheetname]:
        width = sheet_column_widths[sheetname][column]
        writer.sheets[sheetname].set_column(column, width)
```

```python
for sheetname in sheet_percentage_columns:
    columns = sheet_percentage_columns[sheetname]
    for column in columns:
        writer.sheets[sheetname].set_column(column, 12, percentage_format)
```

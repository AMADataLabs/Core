```python
import pandas as pd
```

```python
customer_df = pd.read_csv('Data/TransCustomers.csv')
```

```python
orders_by_email = customer_df.groupby('EMAIL').count()['Customer Number']
prof_by_email = customer_df.groupby('EMAIL').sum()['Total Profiles']
rev_by_email = customer_df.groupby('EMAIL').sum()['Total Rev']
```

```python
orders_by_company = customer_df.groupby('Company Name').count()['Customer Number']
rev_by_company = customer_df.groupby('Company Name').sum()['Total Rev']
prof_by_company = customer_df.groupby('Company Name').sum()['Total Profiles']
```

```python
orders_by_date = customer_df.groupby('Order Date').count()['Customer Number']
rev_by_date = customer_df.groupby('Order Date').sum()['Total Rev']
prof_by_date = customer_df.groupby('Order Date').sum()['Total Profiles']
```

```python
email = pd.DataFrame([orders_by_email, prof_by_email, rev_by_email]).transpose().rename(columns={'Customer Number':'Total Orders'})
company = pd.DataFrame([orders_by_company, prof_by_company, rev_by_company]).transpose().rename(columns={'Customer Number':'Total Orders'})
date = pd.DataFrame([orders_by_date, prof_by_date, rev_by_date]).transpose().rename(columns={'Customer Number':'Total Orders'})
```

```python
company_deets = customer_df[['Company Name','Isell Login','Customer Number','Customer Category']].drop_duplicates()
email_deets = customer_df[['EMAIL','Isell Login','Customer Number','Company Name','Customer Category','FIRST_NM','LAST_NM']].drop_duplicates()
```

```python
company_df = pd.merge(company_deets, company, on='Company Name')
email_df = pd.merge(email_deets, email, on='EMAIL')
```

```python
date
```

```python
with pd.ExcelWriter('GoogleMarketingActivity.xlsx') as writer:  # doctest: +SKIP
    company_df.to_excel(writer, sheet_name='By Company',index=False)
    email_df.to_excel(writer, sheet_name='By Email',index=False)
    date.to_excel(writer, sheet_name='By Date')
```

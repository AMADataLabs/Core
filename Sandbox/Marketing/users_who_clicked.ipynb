```python
import pandas as pd
from fuzzywuzzy import fuzz
```

```python
users = pd.read_csv('../../../Data/Martketing/users_who_clicked.csv')
jan = pd.read_excel('../../../Data/Martketing/jan.xlsx')
dec = pd.read_excel('../../../Data/Martketing/dec.xlsx')
```

```python
users = users.fillna('None')
jan = jan.fillna('None')
dec = dec.fillna('None')
```

```python
user_comp = list(users.drop_duplicates('company1')['company1'])
```

```python
# jan_comp = list(jan['Company Name'])
# dec_comp = list(dec['Company Name'])
# trans = jan_comp+dec_comp
trans = [v.upper() for v in trans_comp]
user = [z.upper() for z in user_comp]
```

```python
dictlist =[]
for company in user:
    score = 0
    print('----')
    print(company)
    for company_2 in trans:
        company2 = 'None'
        if company == company_2:
            print(f'{company} exactly matched to {company_2}!')
            company2 = company_2
            score = 100
            new_dict = {'Company_Campaign':company,
                       'Company_Purchase': company2,
                       'Score':score}
            dictlist.append(new_dict)
            break
        else:
            rate = fuzz.ratio(company, company_2)
            if company in company_2:
                print(f'{company} partially matched to {company_2}! {rate}')
                company2 = company_2
                score = rate + 10
            elif company_2 in company:
                print(f'{company} partially  matched to {company_2}! {rate}')
                company2 = company_2
                score = rate + 10
            elif rate >  60:
                print(f'{company} potentially  matched to {company_2}! {rate}')
                company2 = company_2
                score = rate
        if company2 != 'None':
            new_dict = {'Company_Campaign':company,
                       'Company_Purchase': company2,
                       'Score':score}
            dictlist.append(new_dict)
```

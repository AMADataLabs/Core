```python
all_data['Advantage_Number']=all_data['Advantage Customer #']
```

```python
all_data = all_data.rename(columns={'Total Rev': 'Total_Rev'})
```

```python
company_dict = {}
for company in companies:
    company_dict[company]={}
    company_dict[company]['Revenue Before Campaign']=0
    company_dict[company]['Orders Before Campaign']=0
    company_dict[company]['Revenue After Campaign']=0
    company_dict[company]['Orders After Campaign']=0
```

```python
advantage_dict ={}
for num in numbers:
    advantage_dict[num]={}
    advantage_dict[num]['Revenue Before Campaign']=0
    advantage_dict[num]['Orders Before Campaign']=0
    advantage_dict[num]['Revenue After Campaign']=0
    advantage_dict[num]['Orders After Campaign']=0
```

```python
for row in all_data.itertuples():
    num = float(row.Advantage_Number)
    if num in advantage_dict.keys():
        if row.YEAR < 2019:
            advantage_dict[num]['Revenue Before Campaign']+=row.Total_Rev
            advantage_dict[num]['Orders Before Campaign']+=1
        elif row.YEAR == 2019:
            if row.MonthName in ['August','September','October','November','December']:
                advantage_dict[num]['Revenue After Campaign']+=row.Total_Rev
                advantage_dict[num]['Orders After Campaign']+=1
            else:
                advantage_dict[num]['Revenue Before Campaign']+=row.Total_Rev
                advantage_dict[num]['Orders Before Campaign']+=1
```

```python
advantage_df = pd.DataFrame(advantage_dict).transpose().sort_values('Revenue After Campaign', ascending=False)
```

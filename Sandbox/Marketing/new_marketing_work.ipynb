```python
import pandas as pd
from fuzzywuzzy import fuzz
```

```python
users = pd.read_csv('../../../Data/Martketing/users_who_clicked.csv')
jan = pd.read_excel('../../../Data/Martketing/jan.xlsx')
dec = pd.read_excel('../../../Data/Martketing/dec.xlsx')
match = pd.read_csv('../../../Data/Martketing/company_matches_2.csv')
```

```python
users=users.fillna('None')
jan=jan.fillna('None')
dec=dec.fillna('None')
users.company1 = [v.upper() for v in users.company1]
jan['Company Name'] = [w.upper() for w in jan['Company Name']]
dec['Company Name'] = [x.upper() for x in dec['Company Name']]
all_purchase = pd.concat([jan, dec])
```

```python
users_ = pd.merge(users, match, left_on = 'company1', right_on='Company_Campaign')
all_ = pd.merge(all_purchase, match, left_on = 'Company Name', right_on='Company_Purchase')
```

```python
all_2 = pd.merge(users_, all_, on = 'Company_Campaign')
```

```python
all_2.drop_duplicates(['MONTH','company1']).groupby('company1').sum().to_csv('../../../Data/Martketing/Email_Journey_Users_Purchasing_20-03-04.csv')
```

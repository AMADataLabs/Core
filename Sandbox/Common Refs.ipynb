```python
import pandas
```

```python
refs_common_code = pandas.read_csv('refs_common_code.csv')
refs_common_code['Project'] = refs_common_code['File'].str.split('/').apply(lambda x: x[1])
refs_common_code['Filename'] = refs_common_code['File'].str.split('/').apply(lambda x: x[2])
# refs_common_code = refs_common_code.drop('File', axis=1)
refs_common_code
```

```python
refs_common_model_code = pandas.read_csv('refs_common_model_code.csv')
refs_common_model_code['Project'] = refs_common_model_code['File'].str.split('/').apply(lambda x: x[1])
refs_common_model_code['Filename'] = refs_common_model_code['File'].str.split('/').apply(lambda x: x[2])
# refs_common_model_code = refs_common_model_code.drop('File', axis=1)
refs_common_model_code
```

```python
print('Common Code')
for module in sorted(refs_common_code['Module'].unique()):
    print(f'* *{module}* - ')
print('')
print('Common Model Code')
for module in sorted(refs_common_model_code['Module'].unique()):
    print(f'* *{module}* - ')
```

```python
for module in sorted(refs_common_code['Module'].unique()):
    print(f'Files using CommonCode module {module}')

    for file in refs_common_code[refs_common_code['Module'] == module]['File'].tolist():
        print(f'\t{file}')

print('===========================================================================================================')

for module in sorted(refs_common_model_code['Module'].unique()):
    print(f'Files using CommonModuleCode module {module}')

    for file in refs_common_model_code[refs_common_model_code['Module'] == module]['File'].tolist():
        print(f'\t{file}')
```

```python
for module in sorted(refs_common_code['Module'].unique()):
    print(f'Projects using CommonCode module {module}')

    for project in refs_common_code[refs_common_code['Module'] == module]['Project'].unique().tolist():
        print(f'\t{project}')

print('===========================================================================================================')

for module in sorted(refs_common_model_code['Module'].unique()):
    print(f'Projects using CommonModuleCode module {module}')

    for project in refs_common_model_code[refs_common_model_code['Module'] == module]['Project'].unique().tolist():
        print(f'\t{project}')
```

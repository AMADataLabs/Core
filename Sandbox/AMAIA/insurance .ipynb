```python
import pandas as pd
```

```python
wslive_data = pd.read_csv("../Phone API Work/connected.csv")
insurance_updates_2019 = pd.read_csv("../address_updates_2019.csv")
```

```python
insurance_df = pd.merge(wslive_data, insurance_updates_2019, left_on="PHYSICIAN_ME_NUMBER", right_on="ME")
```

```python
insurance_df
```

```python
insurance_df = insurance_df.dropna(subset=['ADDR_1'])
dict_list =[]
count = 0
total = 0
for row in insurance_df.itertuples():
    total += 1
    new_dict = {}
    address_1 = str(row.OFFICE_ADDRESS_LINE_2).lower()
    address_2 = str(row.ADDR_1).lower()
    address_1_long = str(f'{row.OFFICE_ADDRESS_LINE_2} {row.OFFICE_ADDRESS_LINE_1}').lower()
    address_2_long = str(f'{row.ADDR_1} {row.ADDR_2}').lower()
    if address_2 not in address_1 and address_1 not in address_2_long:
        if longify(address_1) not in longify(address_2):
            if longify(str(row.OFFICE_ADDRESS_CITY)).lower() != longify(str(row.CITY)).lower():
                count += 1
                print(f'{row.OFFICE_ADDRESS_CITY} is not {row.CITY}')
                new_dict["Address_MF"] = row.OFFICE_ADDRESS_LINE_2
                new_dict["City_MF"] = row.OFFICE_ADDRESS_CITY
                new_dict["Address_IA"] = row.ADDR_1
                new_dict["City_IA"] = row.CITY
                new_dict["ME"] = row.PHYSICIAN_ME_NUMBER
                new_dict["Name"] = f'{row.PHYSICIAN_FIRST_NAME} {row.PHYSICIAN_LAST_NAME}'
                new_dict['Date'] = str(row.FROM_DT)[0:10]
                dict_list.append(new_dict)
updated = pd.DataFrame(dict_list)
print (f'{count/total * 100}% updated')
updated
```

```python
def longify(address):
    place = address.replace(" n "," north ")
    place = place.replace(" s "," south ")
    place = place.replace(" e "," east ")
    place = place.replace(" w "," west ")
#     place = place.replace(" Dr."," drive")
#     place = place.replace(" Ave"," avenue")
#     place = place.replace(" Pl"," place")
    place = place.replace(" ln", " lane")
    place = place.replace(" ct", " court")
    place = place.replace(" rd", " road")
    place = place.replace("  ", " ")
    
    return(place)
```

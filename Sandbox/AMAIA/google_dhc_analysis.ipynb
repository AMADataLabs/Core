```python
import pandas as pd
from fuzzywuzzy import fuzz
```

```python
sample = pd.read_csv('../../../Data/AMAIA/True_Sample.csv')
```

```python
goog = pd.read_csv('../../../Data/AMAIA/Google_Insurance_2.csv')
```

```python
google = goog[['TELEPHONE_NUMBER', 'Name', 'Address', 'Types', 'Phone_Number_mar']]
```

```python
test = pd.merge(sample, google, on='Phone_Number_mar')
```

```python
test = test.rename(columns={'Physician Name_mar':'Physician_Name_mar'})
```

```python
all_me = list(sample['ME'])
```

```python
dict_list = []
for me in all_me:
    name_score = 0
    new_df = test[test.ME==me]
    for row in new_df.itertuples():
        score = fuzz.ratio(row.Name, row.Physician_Name_mar)
        if score>name_score:
            name_score=score
            NAME = row.Name
    print(f'{row.Physician_Name_mar} matched to {NAME}')
    new_dict = {
        'ME':me,
        'NAME':NAME,
        'SCORE':name_score
    }
    dict_list.append(new_dict)
```

```python
test = test.rename(columns = {'Last Name_mar':'Last_Name_mar'})
```

```python
scores =[]
for row in test.itertuples():
    goog_add = row.Address.split(',')
    name_score = fuzz.ratio(row.Name, row.Physician_Name_mar)
    if row.Last_Name_mar in row.Name:
        name_score += 100
    
    add_score = fuzz.ratio(row.ADDR_1,goog_add[0])
    if len(goog_add)>1:
        city_score = fuzz.ratio(row.CITY,goog_add[1])
        zip_score = fuzz.ratio(str(row.ZIP), str(goog_add[2].split(' ')[0]))
        score = add_score + name_score + city_score + zip_score
    else:
        score = add_score + name_score
    scores.append(score)
```

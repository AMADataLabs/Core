```python
#Import dependencies
import pandas as pd
```

```python
#Read CSVs of POLO and PPMA data
PPMA = pd.read_csv('PPMA_DF.csv')
POLO = pd.read_csv('POLO_DF.csv')
```

```python
#Merge dataframes
merged = pd.merge(PPMA, POLO, on = 'ME', suffixes = ("_PPMA", "_POLO"))
```

```python
#Define summarizing function:
def sum_it_up(df):
    count =0 
    this_count =0
    dict_list = []
    date_count = 0
    total = df.shape[0]
    for row in df.itertuples():
        date_mf = pd.to_datetime(row.DATE_masterfile)
        date_in = pd.to_datetime(row.DATE_insurance)
        if row.STATE_insurance != row.STATE_masterfile and row.ADDR_1_insurance != row.ADDR_1_masterfile and str(row.CITY_insurance).lower() not in str(row.CITY_masterfile).lower() and str(row.ZIP_insurance) != str(row.ZIP_masterfile) and  date_in > date_mf:
            this_count += 1
            new_dict = {}
            new_dict['ME'] = row.ME
            new_dict['ADDR_1_insurance'] = row.ADDR_1_insurance
            new_dict['ADDR_1_masterfile'] = row.ADDR_1_masterfile
            new_dict['ADDR_2_insurance'] = row.ADDR_2_insurance
            new_dict['ADDR_2_masterfile'] = row.ADDR_2_masterfile
            new_dict['ADDR_3_insurance'] = row.ADDR_3_insurance
            new_dict['ADDR_3_masterfile'] = row.ADDR_3_masterfile
            new_dict['ZIP_insurance'] = row.ZIP_insurance
            new_dict['ZIP_masterfile'] = row.ZIP_masterfile
            new_dict['CITY_insurance'] = row.CITY_insurance
            new_dict['CITY_masterfile'] = row.CITY_masterfile
            new_dict['DATE_insurance'] = row.DATE_insurance
            new_dict['DATE_masterfile'] = row.DATE_masterfile
            new_dict['STATE_insurance'] = row.STATE_insurance
            new_dict['STATE_masterfile'] = row.STATE_masterfile
            dict_list.append(new_dict)


    print(f'''
    {this_count/total * 100}% insurance addresses do not match masterfile 
    ''')
    return(dict_list)
```

```python
#Define smaller dataframe of only extreme mismatches
polo_updates = sum_it_up(POLO)
```

```python
#Define smaller dataframe of only extreme mismatches
pma_updates = sum_it_up(PPMA)
```

```python
#Write results to dataframes
POLO_DF = pd.DataFrame(polo_updates)
PPMA_DF = pd.DataFrame(pma_updates)
```

```python
#Merge new dataframes
X = pd.merge(POLO_DF, PPMA_DF, on='ME', suffixes =('_POLO', '_PPMA'))
X = X.drop(columns=['ADDR_1_insurance_POLO','ADDR_2_insurance_POLO','ADDR_3_insurance_POLO', 'DATE_insurance_POLO', 'CITY_insurance_POLO', 'ZIP_insurance_POLO', 'STATE_insurance_POLO'])
X = X.rename(columns={'ADDR_1_insurance_PPMA':'ADDR_1_insurance','ADDR_2_insurance_PPPMA': 'ADDR_2_insurance','ADDR_3_insurance_PPMA': 'ADDR_3_insurance', 'DATE_insurance_PPMA': 'DATE_insurance', 'CITY_insurance_PPMA':'CITY_insurance', 'ZIP_insurance_PPMA':'ZIP_insurance', 'STATE_insurance_POLO':"STATE_insurance"})
X.to_csv('polo_ppma_no_matches.csv')
```
